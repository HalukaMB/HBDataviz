```python
tickers_dict = {
 'Mexico':"D5BI.DE",
 'Türkiye, Republic of':"IQQ5.DE",
  'Vietnam':"XFVT.DE",
 "Hungary":"FHUBE.DE",
 "Qatar":"XGLF.DE",
 "Serbia":"ESNB.SG"}
```

```python
data_dict={}
for country, ticker in tickers_dict.items():
    try:
        data = yfinance.download(ticker, start="2019-01-01")
        if not data.empty:
            print(f"{country} ({ticker}):")
            print(data['Close'].pct_change().dropna().head())
            data_dict[country] = data['Close'].dropna()
    except Exception as e:
        print(f"Error fetching data for {country} ({ticker}): {e}")
```

```python
((data_dict["Türkiye, Republic of"]-data_dict["Türkiye, Republic of"].shift(100))/data_dict["Türkiye, Republic of"].shift(100)).min()
```

```python
def calculate_change(shift_days):
    array_of_countries=[]
    for key in data_dict.keys():
        name_of_country=key
        max_change=(100*(data_dict[name_of_country]-data_dict[name_of_country].shift(shift_days))/data_dict[name_of_country].shift(shift_days)).max().values[0]
        min_change=(100*(data_dict[name_of_country]-data_dict[name_of_country].shift(shift_days))/data_dict[name_of_country].shift(shift_days)).min().values[0]
        median_change=(100*(data_dict[name_of_country]-data_dict[name_of_country].shift(shift_days))/data_dict[name_of_country].shift(shift_days)).median().values[0]

        row=[name_of_country,max_change,min_change,median_change]
        array_of_countries.append(row)
    df=pd.DataFrame(array_of_countries, columns=["Land", "Maximum", "Minimum","Median"])
    df["Tage"]=shift_days
    return df
```

```python
array_of_days=[50,100,200]
df=pd.DataFrame(columns=["Land", "Maximum", "Minimum", "Median", "Tage"])
for days in array_of_days:
    calculate_change(days)
    df=pd.concat([df,calculate_change(days)])
df
```
