```python
alcohol_ban=pd.read_csv("../data/Alcohol_ban.csv", index_col=0)
alcohol_ban
```

```python
alcohol_ban["ban"]=1
alcohol_ban.reset_index(inplace=True)
alcohol_ban
```

```python
country_to_iso3 = {}
for name in alcohol_ban.Name.tolist():
    lookup_name = name.replace("*", "").strip()
    if name in overrides:
        country_to_iso3[name] = overrides[name]
    else:
        try:
            country_to_iso3[name] = pycountry.countries.lookup(lookup_name).alpha_3
        except LookupError:
            country_to_iso3[name] = None
country_to_iso3
```

```python
alcohol_ban["Country Code"] = alcohol_ban["Name"].map(country_to_iso3)
alcohol_ban
```

```python
alcohol_ban_red=alcohol_ban[["Country Code","ban"]]
alcohol_ban_red
```
