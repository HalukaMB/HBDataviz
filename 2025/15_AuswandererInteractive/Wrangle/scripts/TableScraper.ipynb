```python
import pycountry

def get_iso3(name):
    # Handle some common exceptions
    overrides = {
        "Czechia": "CZE",
        "Democratic Republic of the Congo": "COD",
        "Republic of the Congo": "COG",
        "Eswatini": "SWZ",
        "North Macedonia": "MKD",
        "Palestine": "PSE",
        "South Korea": "KOR",
        "North Korea": "PRK",
        "Vatican City": "VAT",
        "Taiwan": "TWN",
        "United States": "USA",
        "United Kingdom": "GBR",
        "Russia": "RUS",
        "Syria": "SYR",
        "Laos": "LAO",
        "Moldova": "MDA",
        "Bolivia": "BOL",
        "Tanzania": "TZA",
        "Venezuela": "VEN",
        "Brunei": "BRN",
        "Ivory Coast": "CIV",
        "Cape Verde": "CPV",
        "East Timor": "TLS",
        "Micronesia": "FSM",
        "Saint Kitts and Nevis": "KNA",
        "Saint Lucia": "LCA",
        "Saint Vincent and the Grenadines": "VCT",
        "Sao Tome and Principe": "STP",
        "Solomon Islands": "SLB",
        "Trinidad and Tobago": "TTO",
        "United Arab Emirates": "ARE",
        "Antigua and Barbuda": "ATG",
        "Bahamas": "BHS",
        "Gambia": "GMB",
        "Guinea-Bissau": "GNB",
        "Marshall Islands": "MHL",
        "Myanmar": "MMR",
        "San Marino": "SMR",
        "South Sudan": "SSD",
        "Suriname": "SUR",
        "Tuvalu": "TUV",
        "Vanuatu": "VUT",
        "Kosovo": "XKX",
        "Timor-Leste": "TLS",
        "Cabo Verde": "CPV",
    }
    if name in overrides:
        return overrides[name]
    try:
        return pycountry.countries.lookup(name).alpha_3
    except LookupError:
        return None

df_timezones["ISO3"] = df_timezones["Country"].apply(get_iso3)
df_timezones
```

```python
# Extract the first number from "Anzahl der Deutschen" column
import re

def extract_first_number(text):
    """
    Extract the first number from a string, handling German number formatting
    """
    if pd.isna(text):
        return None
    
    # Convert to string if not already
    text = str(text)
    
    # Remove common German thousand separators and replace decimal comma with dot
    text = text.replace('.', '').replace(',', '.')
    
    # Find the first number (including decimals)
    match = re.search(r'\d+(?:\.\d+)?', text)
    
    if match:
        try:
            return float(match.group())
        except ValueError:
            return None
    return None

# Apply the function to extract first numbers
df_red["Anzahl_der_Deutschen_numeric"] = df_red["Anzahl der Deutschen"].apply(extract_first_number)

# Show the results
print("Original vs cleaned numbers:")
print(df_red[["Anzahl der Deutschen", "Anzahl_der_Deutschen_numeric"]].head(10))
```
