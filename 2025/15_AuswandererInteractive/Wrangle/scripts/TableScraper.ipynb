```python
import pandas as pd
import glob
```

```python
files=glob.glob('../data/*/API*.csv')
files
```

```python
gdp_data=pd.read_csv("../data/API_NY.GDP.PCAP.CD_DS2_en_csv_v2_104/API_NY.GDP.PCAP.CD_DS2_en_csv_v2_104.csv",skiprows=3)
gdp_data
```

```python
gdp_data = gdp_data.fillna(method='ffill', axis=1)
gdp_data
```

```python
us_gdp=pd.to_numeric(gdp_data[gdp_data["Country Code"]=="USA"]["2024"],errors="coerce")
```

```python
gdp_data["recent"]=pd.to_numeric(gdp_data["2024"],errors="coerce")
gdp_data
```

```python
gdp_data["recent"]
```

```python
gdp_data["gdp_rel_us"]=gdp_data["recent"]/us_gdp.values[0]
gdp_data
```

```python
files[3]
```

```python
ppp_data=pd.read_csv("../data/API_PA.NUS.PPPC.RF_DS2_en_csv_v2_563/API_PA.NUS.PPPC.RF_DS2_en_csv_v2_563.csv",skiprows=3)
ppp_data = ppp_data.fillna(method='ffill', axis=1)
ppp_data["ppp_rel_us"]=pd.to_numeric(ppp_data["2024"],errors="coerce")
ppp_data=ppp_data.dropna(subset=["ppp_rel_us"])
ppp_data
```

```python
df_rel_us=pd.merge(ppp_data[["Country Name","Country Code","ppp_rel_us"]],gdp_data[["Country Code","gdp_rel_us"]],on="Country Code")
df_rel_us
```

```python
df_rel_us["cost_to_gdp"]=df_rel_us["ppp_rel_us"]/df_rel_us["gdp_rel_us"]
df_rel_us
```

```python
df_rel_us.sort_values("cost_to_gdp",ascending=False,inplace=True)
df_rel_us=df_rel_us.reset_index(drop=True)
df_rel_us.tail(10)
```

```python
df_ppp_gdp_cost=df_rel_us.copy()
```

```python
popdensity=pd.read_csv( '../data/API_EN.POP.DNST_DS2_en_csv_v2_5836/API_EN.POP.DNST_DS2_en_csv_v2_5836.csv',skiprows=3)
popdensity = popdensity.fillna(method='ffill', axis=1)
popdensity["popdensity"]=pd.to_numeric(popdensity["2024"],errors="coerce")
popdensity_df=popdensity[["Country Code","popdensity"]]
popdensity_df
```

```python
df_general=pd.merge(df_ppp_gdp_cost,popdensity_df,on="Country Code")
df_general
```
