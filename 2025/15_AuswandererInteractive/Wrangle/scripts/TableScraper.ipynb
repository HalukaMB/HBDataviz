```python
import pycountry

def get_iso3(name):
    # Handle some common exceptions
    overrides = {
        "Czechia": "CZE",
        "Democratic Republic of the Congo": "COD",
        "Republic of the Congo": "COG",
        "Eswatini": "SWZ",
        "North Macedonia": "MKD",
        "Palestine": "PSE",
        "South Korea": "KOR",
        "North Korea": "PRK",
        "Vatican City": "VAT",
        "Taiwan": "TWN",
        "United States": "USA",
        "United Kingdom": "GBR",
        "Russia": "RUS",
        "Syria": "SYR",
        "Laos": "LAO",
        "Moldova": "MDA",
        "Bolivia": "BOL",
        "Tanzania": "TZA",
        "Venezuela": "VEN",
        "Brunei": "BRN",
        "Ivory Coast": "CIV",
        "Cape Verde": "CPV",
        "East Timor": "TLS",
        "Micronesia": "FSM",
        "Saint Kitts and Nevis": "KNA",
        "Saint Lucia": "LCA",
        "Saint Vincent and the Grenadines": "VCT",
        "Sao Tome and Principe": "STP",
        "Solomon Islands": "SLB",
        "Trinidad and Tobago": "TTO",
        "United Arab Emirates": "ARE",
        "Antigua and Barbuda": "ATG",
        "Bahamas": "BHS",
        "Gambia": "GMB",
        "Guinea-Bissau": "GNB",
        "Marshall Islands": "MHL",
        "Myanmar": "MMR",
        "San Marino": "SMR",
        "South Sudan": "SSD",
        "Suriname": "SUR",
        "Tuvalu": "TUV",
        "Vanuatu": "VUT",
        "Kosovo": "XKX",
        "Timor-Leste": "TLS",
        "Cabo Verde": "CPV",
    }
    if name in overrides:
        return overrides[name]
    try:
        return pycountry.countries.lookup(name).alpha_3
    except LookupError:
        return None

df_timezones["ISO3"] = df_timezones["Country"].apply(get_iso3)
df_timezones
```

```python
url_religion="https://en.wikipedia.org/wiki/Religions_by_country"
```

```python
response = requests.get(url_religion)

soup = BeautifulSoup(response.text, "html.parser")

# Find the first table on the page
table = soup.find_all("table", class_="wikitable")

# Read the HTML table into a DataFrame
```

```python
base_df=pd.DataFrame()
for i in range(1,len(table)):
    df = pd.read_html(str(table))[i]
    alt_cols=[]
    for i in df.columns:
        alt_cols.append(i[0]+"_"+i[1])
    df.columns=alt_cols
    base_df=pd.concat([base_df,df], ignore_index=True)
base_df
```

```python
base_df.to_csv("../data/religion.csv", index=False)
```
