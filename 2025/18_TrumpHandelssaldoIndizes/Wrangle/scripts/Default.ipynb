```python
import pandas as pd
import glob
```

```python
files=glob.glob('Countries_Yahoo/*.csv')
files
```

```python
data_list = []
for i in range(len(files)):

    country=files[i][16:-10]
    data=pd.read_csv(files[i])
    data=data.iloc[2:-2]
    data["Close"]=data["Close"].astype(float)
    gain_since_trump=(data.iloc[-1]["Close"]/data.iloc[0]["Close"]-1)*100
    fluctuation=(data["Close"].max()/data["Close"].min()-1)*100
    row_dict={"country": country,
    "gain_since_trump": gain_since_trump,
    "fluctuation": fluctuation,
    "min": data["Close"].min(),
    "max": data["Close"].max()}
    data_list.append(row_dict)
yahoo_df=pd.DataFrame(data_list)
```

```python
files=glob.glob('Countries_Investing/*.csv')
files
```

```python
data=pd.read_csv(files[0])
data
```

```python
data_list = []
for i in range(len(files)):

    country=files[i][20:-14]
    data=pd.read_csv(files[i])
    data["Date"]=pd.to_datetime(data["Date"], format="%m/%d/%Y")
    data=data.sort_values(by="Date", ascending=True).reset_index(drop=True)
    data["Close"]=data["Price"].str.replace(",","").astype(float)
    gain_since_trump=(data.iloc[-1]["Close"]/data.iloc[0]["Close"]-1)*100
    fluctuation=(data["Close"].max()/data["Close"].min()-1)*100
    row_dict={"country": country,
    "gain_since_trump": gain_since_trump,
    "fluctuation": fluctuation,
    "min": data["Close"].min(),
    "max": data["Close"].max()}
    data_list.append(row_dict)
investing_df=pd.DataFrame(data_list)
```

```python
df_indices=pd.concat([yahoo_df, investing_df], ignore_index=True)
df_indices
```

```python
df_indices.sort_values(by="gain_since_trump", ascending=False, inplace=True)
df_indices
```

```python
df_indices.to_csv("indices.csv", index=False)
```
