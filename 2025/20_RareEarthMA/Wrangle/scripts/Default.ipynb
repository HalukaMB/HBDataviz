```python
import pandas as pd
import glob
```

```python
files=glob.glob('../data/*LSEG.xlsx')
files
```

```python
df=pd.DataFrame()
for i in range(len(files)):
    countryname=files[i][11:-15]
    print(countryname)
    data=pd.read_excel(files[i])
    data["country"]=countryname
    df=pd.concat([df, data], ignore_index=True)
df
```

```python
red_df=df[["country","Date Announced","Rank Value inc. Net Debt of Target\n(USD, Millions)","Target Nation","Deal Status"]]
red_df
```

```python
red_df["Year"]=red_df["Date Announced"].astype(str).str[:4]
red_df["Year"]=red_df["Year"].astype(int)
red_df
```

```python
red_df["Value"]=pd.to_numeric(red_df["Rank Value inc. Net Debt of Target\n(USD, Millions)"],errors="coerce")
red_df
```

```python
announced_sums_by_country=red_df.groupby("country")["Value"].sum().sort_values(ascending=False).reset_index()
announced_sums_by_country
```

```python
ma_undisclosed_sums_by_country=red_df[red_df["Value"].isna()].groupby("country")["Year"].count().sort_values(ascending=False).reset_index()
ma_undisclosed_sums_by_country
```

```python
mas_overall=red_df.groupby("country")["Value"].count().sort_values(ascending=False).reset_index()

mas_overall
```

```python
table_of_target_nations=red_df.groupby(["country","Target Nation"])["Value"].count().reset_index()
table_of_target_nations
```

```python
target_nations_by_sum=red_df.groupby(["country","Target Nation"])["Value"].sum().reset_index()
target_nations_by_sum.sort_values(by="Target Nation", ascending=True, inplace=True)
target_nations_by_sum
```
