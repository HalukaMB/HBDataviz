```python
import pandas as pd
import glob
```

```python
files=glob.glob('../data/*LSEG_red.xlsx')
files
```

```python
df=pd.DataFrame()
for i in range(len(files)):
    countryname=files[i][11:-19]
    print(countryname)
    data=pd.read_excel(files[i])
    data["country"]=countryname
    df=pd.concat([df, data], ignore_index=True)
df
```

```python
red_df=df[["country","Date Announced","Rank Value inc. Net Debt of Target\n(USD, Millions)","Target Nation","Deal Status"]]
red_df
```

```python
red_df["Year"]=red_df["Date Announced"].astype(str).str[:4]
red_df["Year"]=red_df["Year"].astype(int)
red_df
```

```python
red_df["Value"]=pd.to_numeric(red_df["Rank Value inc. Net Debt of Target\n(USD, Millions)"],errors="coerce")
red_df
```

```python
announced_sums_by_country=red_df.groupby("country")["Value"].sum().sort_values(ascending=False).reset_index()
announced_sums_by_country["Value_in_Mrd"]=announced_sums_by_country["Value"]/1000
announced_sums_by_country
```

```python
announced_sums_by_country.to_csv("sums_by_three.csv", index=False)
```

```python
mas_overall=red_df.groupby("country")["Year"].count().sort_values(ascending=False).reset_index()

mas_overall
```

```python
mas_overall_year=red_df.groupby(["country","Year"])["Date Announced"].count().sort_values(ascending=False).reset_index()

mas_overall_year
```

```python
mas_overall_year_pivot=mas_overall_year.pivot(index="Year", columns="country", values="Date Announced").fillna(0).astype(int).reset_index()
mas_overall_year_pivot[["Year","China","USA","EU"]]
```

```python
mas_overall_year_pivot[["Year","China","USA","EU"]].to_csv("number_investments_by_three_by_year.csv", index=False)
```

```python
ma_undisclosed_sums_by_country=red_df[red_df["Value"].isna()].groupby("country")["Year"].count().sort_values(ascending=False).reset_index()
ma_undisclosed_sums_by_country
```

```python
overall_vs_undisclosed_ma=pd.merge(mas_overall, ma_undisclosed_sums_by_country, on="country", how="inner")
overall_vs_undisclosed_ma
```

```python
overall_vs_undisclosed_ma["Known"]=overall_vs_undisclosed_ma["Year_x"]-overall_vs_undisclosed_ma["Year_y"]
overall_vs_undisclosed_ma
```

```python
overall_vs_undisclosed_ma=overall_vs_undisclosed_ma.rename(columns={"Year_x":"Overall MA Count","Known":"Bekannt" ,"Year_y":"Unbekannt"})
overall_vs_undisclosed_ma=overall_vs_undisclosed_ma[["country","Bekannt","Unbekannt","Overall MA Count"]]
overall_vs_undisclosed_ma
```

```python
overall_vs_undisclosed_ma.to_csv("number_investments_by_three.csv", index=False)
```

```python
table_of_target_nations=red_df.groupby(["country","Target Nation"])["Year"].count().reset_index()
table_of_target_nations
```

```python
table_of_target_nations_pivot=table_of_target_nations.pivot(columns="country", index="Target Nation", values="Year").fillna(0).astype(int).reset_index()
table_of_target_nations_pivot
```

```python
table_of_target_nations_pivot.to_csv("table_of_target_nations.csv", index=False)
```
