```python
# Fetch data for each stock by ticker
def get_stock_data_dates(ticker, start_date, end_date=None):
    try:
        stock = yf.Ticker(ticker)
        return stock.history(start=start_date, end=end_date)
    except:
        print("Ticker")
        print("Not found")
        return None

#  Get data from start of 2000 to end of 2025
start_date = "2000-01-01"
end_date = "2025-12-31"


# Get data for each brand
for index, row in luxury_brands.iterrows():
    print(row["Company Name"])
    print(row['Rank'])
    ticker = row['ticker']
    print(f"Getting data for {ticker}")
    data = get_stock_data_dates(ticker, start_date, end_date)
    # If there is data, we write the Close prices into a column named after the company
    if data is not None:
        print(f"Latest close price: {data['Close'].iloc[-1]:.2f}")
        data[row["Company Name"]]= data['Close']
        #We reset the index to have Date as a column
        data.reset_index(inplace=True)
        data["Date"] = pd.to_datetime(data["Date"]).dt.date
        #We reduce the dataframe to Date and Company Name only
        red_data=data[["Date",row["Company Name"]]]
        #if it is the first iteration, we create the base dataframe
        if index==0:
            base_df=red_data
        #else we merge the new data into the base dataframe
        else:
            base_df=pd.merge(base_df,red_data,on="Date",how="outer")
```

```python
#we cut of the last row with incomplete data
red_base_df=base_df.iloc[:-1]
red_base_df
```

```python
#and then we get the S&P500 data for the same time frame
sandp500=yf.Ticker("^GSPC")
sp500_data=sandp500.history(start=start_date, end=end_date)
sp500_data.reset_index(inplace=True)
sp500_data["Date"] = pd.to_datetime(sp500_data["Date"]).dt.date
sp500_data=sp500_data[["Date","Close"]]
sp500_data.rename(columns={"Close":"S&P500"},inplace=True)
sp500_data
```

```python
#then we define what a setback is by HB standards
ruecksetzer_value=0.9
#and what a recovery would be from this setback
recovery_value=1/ruecksetzer_value
```

```python
# And now we want to get all dates where the S&P500 had a setback of 10% 
# or more from its previous maximum

#we set a default value for the previous maximum
before_max=0
#and an empty dictionary to store the setback data
setback_dict={}

#we loop through the S&P500 data to find setbacks
for index, row in sp500_data.iterrows():
    #get date and value for each row
    date=row["Date"]
    value=row["S&P500"]
    #check if the current value is greater than the previous maximum
    if value>before_max:
        before_max=value
        max_date=date
    #if the current value is less than 90 percent of the previous maximum
    # we record this as a setback
    else:
        if value<before_max*ruecksetzer_value:
            print(f"Date: {date}, Value: {value}, Previous Max: {before_max}")
            #and save it into the setback dictionary
            setback_dict[date]=[value,before_max,max_date]
    
```

```python
#we convert the setback dictionary into a dataframe
setback_df=pd.DataFrame.from_dict(setback_dict,orient="index",columns=["S&P500 Setback Values","Previous Max","Max Date"]).reset_index()
setback_df.rename(columns={"index":"Date"},inplace=True) 
setback_df.sort_values("Date",inplace=True)
setback_df
```

```python
#we get the start dates of each setback period
setback_start=setback_df.drop_duplicates(subset=["Previous Max"],keep="first")
setback_start=setback_start.rename(columns={"Date":"Start Setback Date"})
setback_start=setback_start.reset_index(drop=True)
setback_start
```

```python
#we get the end dates of each setback period

setback_end=setback_df.drop_duplicates(subset=["Previous Max"],keep="last")
setback_end=setback_end.rename(columns={"Date":"End Setback Date"})
setback_end=setback_end.reset_index(drop=True)
setback_end
```

```python
#and we merge the start and end dates, start and end value into one dataframe 
# also including Previous Max and Max Date
setback_start_end=pd.merge(setback_start,setback_end,on=["Previous Max","Max Date"],suffixes=('_Start','_End_Dip'))
setback_start_end
```

```python
#and we resort the dataframe for better readability and logical order
df_setback_resorted=setback_start_end[["Max Date","Previous Max","Start Setback Date","S&P500 Setback Values_Start","End Setback Date","S&P500 Setback Values_End_Dip"]]
df_setback_resorted.columns=["Max Date","Value at Max","Start Setback Date","Value at Start Setback","End Setback Date","Value at End Setback"]
df_setback_resorted
```

```python
# and then we get the recovery dates for each setback period with another run through the dataframe
recoverydict={"Recovered Date":[],"Value At Recovered":[]}
for index, row in df_setback_resorted.iterrows():
    #we take the date of the maximum and its value
    date_max=(row["Max Date"])
    value_max=row["Value at Max"]
    #and then we search for the first date after the maximum date
    # where the S&P500 is greater than or equal to the previous maximum value
    recovered_df=sp500_data[(sp500_data["Date"]>date_max)&(sp500_data["S&P500"]>=value_max)]
    print(recovered_df["Date"].iloc[0])
    #we append the found date and value to the recovery dictionary
    recoverydict["Recovered Date"].append(recovered_df["Date"].iloc[0])
    recoverydict["Value At Recovered"].append(recovered_df["S&P500"].iloc[0])
# and we convert the recovery dictionary into a dataframe
recovery_df=pd.DataFrame(recoverydict)
recovery_df["Recovered Date"]=pd.to_datetime(recovery_df["Recovered Date"]).dt.date
recovery_df
```

# Q1: How long does it take to recover from worst timing investment?
Now I am calculating how fast stocks recovered for luxurious brands, if I invested on the peak moment of the S&P500 into the given stock

```python
setback_sp_500_full
```

```python
# we get only a part of this setback dataframe
time_between_max_and_setback=setback_sp_500_full[["Max Date","Start Setback Date"]].copy()
time_between_max_and_setback
```

```python
# we want to collect the date by which each luxury brand 
# recovered from the value that it had at the time of the S&P500 maximum
luxury_recovery_dates = {}
# we iterate through each luxury brand
for brand in luxury_brands["Company Name"].tolist():
    print(brand)
    #create a reduced dataframe for the brand
    # that only contains Date and the brand's stock price
    brand_data=red_base_df[["Date",brand]]
    recovery_dates=[]
    # we iterate through each setback period
    for index, row in time_between_max_and_setback.iterrows():
        #take the date of the previous S&P500 maximum
        #and set it as the investment date in the brand
        date_invest_max=(row["Max Date"])
        #and set the date of the realization of the setback
        date_crash_realisation=(row["Start Setback Date"])
        #then we get the value of the stock of this brand at the investment date / the date of the S&P500 maximum
        #this is the last entry of all dates before or equal to the investment date
        value_brand_at_max_date=brand_data[brand_data["Date"]<=date_invest_max][brand].values[-1]
        #then we search for all dates after the realization of the setback
        # where the brand's stock price is greater than or equal to the value at the investment date
        recovered_df_brand=brand_data[(brand_data["Date"]>date_crash_realisation)&(brand_data[brand]>=value_brand_at_max_date)]
        #we drop all rows with NaN values for the brand
        recovered_df_brand=recovered_df_brand.dropna(subset=[brand])
        #if there are no entries, we append NaN
        if len(recovered_df_brand)==0:
            date_recovered=np.nan
            recovery_dates.append(date_recovered)
            pass
        #else we take the first date where the brand recovered
        else:
            date_recovered=recovered_df_brand.iloc[0]["Date"]
            recovery_dates.append(date_recovered)
            #and print out this result for control purposes
            print(f"Brand: {brand}, Date Invest Max: {date_invest_max}, Value at Max Date: {value_brand_at_max_date}, Date Recovered: {date_recovered}")
    # and then we append the list of recovery dates for this brand to the dictionary
    luxury_recovery_dates[brand]=recovery_dates
#and create a dataframe from the dictionary    
luxury_recovery_df=pd.DataFrame(luxury_recovery_dates)
luxury_recovery_df
```

```python
# we add a column for the recovery dates of the S&P500
df_recovery_concat=pd.concat([setback_sp_500_full[["Max Date","Start Setback Date","Recovered Date"]],luxury_recovery_df],axis=1)
df_recovery_concat
```

```python
# Now we calculate the number of days between the maximum date and the recovery date for the S&P500 and each brand
df_days_recovery=pd.DataFrame()
df_days_recovery["Max Date"]=df_recovery_concat["Max Date"]
df_days_recovery["Start Setback Date"]=df_recovery_concat["Start Setback Date"]

#for each column in the recovery dataframe
for col in df_recovery_concat.columns:
    # if the column is not the Max Date column
    if col not in ['Max Date',"Start Setback Date"]:
        #we calculate the difference in days between the recovery date of said column and the max date
        df_days_recovery[col] =(pd.to_datetime(df_recovery_concat[col]) - pd.to_datetime(df_recovery_concat['Max Date'])).dt.days
```

```python
df_days_recovery
```

```python
#and then we calculate mean and median recovery days across all brands
df_days_recovery["mean_luxury"]=df_days_recovery[df_days_recovery.columns[3:]].mean(axis=1)
df_days_recovery["median_luxury"]=df_days_recovery[df_days_recovery.columns[3:-1]].median(axis=1)
df_days_recovery
```

```python
df_days_recovery['nan_count'] = df_days_recovery[df_days_recovery.columns[3:-1]].isna().sum(axis=1)
df_days_recovery
```

```python
df_days_recovery["Not Recovered"]= df_days_recovery['nan_count']*100/23
```

```python
def get_dividend_data(ticker, start_date, end_date=None):
    try:
        #we get the dividend data
        stock = yf.Ticker(ticker)
        dividends = stock.dividends
        #and cut it down to the requested date range
        if start_date:
            dividends = dividends[dividends.index >= start_date]
        if end_date:
            dividends = dividends[dividends.index <= end_date]
        
        return dividends
    except:
        return None

# Example usage
start_date = "2000-01-01"
end_date = "2024-12-31"

dividend_data = get_dividend_data("MC.PA", start_date, end_date)
print(dividend_data)
```

```python
def check_consecutive_increase(lst):
    skips_list=[]
    for i in range(len(lst) - 1):
        if lst[i+1] != lst[i] + 1:
            skipped_year= lst[i] + 1
            skips_list.append(skipped_year)
    return skips_list
```
