```python
# Fetch data for multiple stocks
def get_stock_data_dates(ticker, start_date, end_date=None):
    try:
        stock = yf.Ticker(ticker)
        return stock.history(start=start_date, end=end_date)
    except:
        return None

# Example: Get data from specific dates
start_date = "2000-01-01"
end_date = "2025-12-31"

base_df=pd.DataFrame()
# Get data for all tickers
for index, row in luxury_brands.iterrows():
    ticker = row['ticker']
    print(f"Getting data for {ticker}")
    data = get_stock_data_dates(ticker, start_date, end_date)
    if data is not None:
        print(f"Latest close price: {data['Close'].iloc[-1]:.2f}")
        data[row["Company Name"]]= data['Close']
        data.reset_index(inplace=True)
        data["Date"] = pd.to_datetime(data["Date"]).dt.date
        red_data=data[["Date",row["Company Name"]]]
        if index==0:
            base_df=red_data
        else:
            base_df=pd.merge(base_df,red_data,on="Date",how="outer")
```

```python
sandp500=yf.Ticker("^GSPC")
sp500_data=sandp500.history(start=start_date, end=end_date)
sp500_data.reset_index(inplace=True)
sp500_data["Date"] = pd.to_datetime(sp500_data["Date"]).dt.date
sp500_data=sp500_data[["Date","Close"]]
sp500_data.rename(columns={"Close":"S&P500"},inplace=True)
```

```python
sp500_data
```

```python
ruecksetzer_value=0.9
recovery_value=1/ruecksetzer_value
```

```python
before_max=0
crash_dict={}
for index, row in sp500_data.iterrows():
    date=row["Date"]
    value=row["S&P500"]
    if value>before_max:
        before_max=value
        max_date=date
    else:
        if value<before_max*ruecksetzer_value:
            print(f"Date: {date}, Value: {value}, Previous Max: {before_max}")
            crash_dict[date]=[value,before_max,max_date]
    
```

```python
crash_df=pd.DataFrame.from_dict(crash_dict,orient="index",columns=["S&P500 Crash Values","Previous Max","Max Date"]).reset_index()
crash_df.rename(columns={"index":"Date"},inplace=True) 
crash_df.sort_values("Date",inplace=True)
crash_df
```

```python
crash_start=crash_df.drop_duplicates(subset=["Previous Max"],keep="first")
crash_start=crash_start.reset_index()
crash_start=crash_start.rename(columns={"index":"Start Crash Date"})
crash_start
```

```python
crash_end=crash_df.drop_duplicates(subset=["Previous Max"],keep="last")
crash_end=crash_end.reset_index()
crash_end=crash_end.rename(columns={"index":"End Crash Date"})
crash_end
```

```python
crash_start_end=pd.merge(crash_start,crash_end,on=["Previous Max","Max Date"],suffixes=('_Start','_End_Dip'))
crash_start_end
```

```python
df_crash_resorted=crash_start_end[["Max Date","Previous Max","Start Crash Date","S&P500 Crash Values_Start","End Crash Date","S&P500 Crash Values_End_Dip"]]
df_crash_resorted.columns=["Max Date","Value at Max","Start Crash Date","Value at Start Crash","End Crash Date","Value at End Crash"]
df_crash_resorted
```

```python
#df=pd.DataFrame({"Date Recovered":[],"Value Recovered":[]})
recoverydict={"Recovered Date":[],"Value At Recovered":[]}
for index, row in df_crash_resorted.iterrows():
    date_max=(row["Max Date"])
    value_max=row["Value at Max"]
    recovered_df=sp500_data[(sp500_data["Date"]>date_max)&(sp500_data["S&P500"]>=value_max)]
    print(recovered_df["Date"].iloc[0])
    recoverydict["Recovered Date"].append(recovered_df["Date"].iloc[0])
    recoverydict["Value At Recovered"].append(recovered_df["S&P500"].iloc[0])
recovery_df=pd.DataFrame(recoverydict)
recovery_df["Recovered Date"]=pd.to_datetime(recovery_df["Recovered Date"]).dt.date
recovery_df
```

```python
crash_sp_500_full=pd.concat([df_crash_resorted,recovery_df],axis=1)
crash_sp_500_full
```

```python
def get_dividend_data(ticker, start_date, end_date=None):
    try:
        stock = yf.Ticker(ticker)
        # Method 1: Using .dividends property
        dividends = stock.dividends
        if start_date:
            dividends = dividends[dividends.index >= start_date]
        if end_date:
            dividends = dividends[dividends.index <= end_date]
        
        return dividends
    except:
        return None

# Example usage
start_date = "2000-01-01"
end_date = "2024-12-31"

dividend_data = get_dividend_data("MC.PA", start_date, end_date)
print(dividend_data)
```

```python
def check_consecutive_increase(lst):
    skips_list=[]
    for i in range(len(lst) - 1):
        if lst[i+1] != lst[i] + 1:
            skipped_year= lst[i] + 1
            skips_list.append(skipped_year)
    return skips_list
```

```python
results_dict={}
for col in base_df.columns[1:]:
    dividends_df=base_df[["Date",col]]
    dividends_df=dividends_df.dropna(subset=[col])
    dividends_df["Date_formatted"] = pd.to_datetime(dividends_df["Date"])
    years_of_dividends=dividends_df["Date_formatted"].dt.year.tolist()
    years_of_dividends=set(years_of_dividends)
    years_of_dividends=sorted(list(years_of_dividends))
    results_dict[col] = check_consecutive_increase(years_of_dividends)
pd.DataFrame.from_dict(results_dict,orient="index",columns=["Skipped Dividends"])
```
