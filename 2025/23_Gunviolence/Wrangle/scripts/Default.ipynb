```python
import pandas as pd
import glob
```

```python
files_mass=glob.glob('../data/massshootingtracker/*.json')
```

```python
dfs = [pd.read_json(f) for f in files_mass]
all_data_mass = pd.concat(dfs, ignore_index=True)
all_data_mass
```

```python
# Dictionary mapping US state abbreviations to full names
us_state_abbrev_to_name = {
    'AL': 'Alabama', 'AK': 'Alaska', 'AZ': 'Arizona', 'AR': 'Arkansas', 'CA': 'California',
    'CO': 'Colorado', 'CT': 'Connecticut', 'DE': 'Delaware', 'FL': 'Florida', 'GA': 'Georgia',
    'HI': 'Hawaii', 'ID': 'Idaho', 'IL': 'Illinois', 'IN': 'Indiana', 'IA': 'Iowa',
    'KS': 'Kansas', 'KY': 'Kentucky', 'LA': 'Louisiana', 'ME': 'Maine', 'MD': 'Maryland',
    'MA': 'Massachusetts', 'MI': 'Michigan', 'MN': 'Minnesota', 'MS': 'Mississippi', 'MO': 'Missouri',
    'MT': 'Montana', 'NE': 'Nebraska', 'NV': 'Nevada', 'NH': 'New Hampshire', 'NJ': 'New Jersey',
    'NM': 'New Mexico', 'NY': 'New York', 'NC': 'North Carolina', 'ND': 'North Dakota', 'OH': 'Ohio',
    'OK': 'Oklahoma', 'OR': 'Oregon', 'PA': 'Pennsylvania', 'RI': 'Rhode Island', 'SC': 'South Carolina',
    'SD': 'South Dakota', 'TN': 'Tennessee', 'TX': 'Texas', 'UT': 'Utah', 'VT': 'Vermont',
    'VA': 'Virginia', 'WA': 'Washington', 'WV': 'West Virginia', 'WI': 'Wisconsin', 'WY': 'Wyoming',
    'DC': 'District of Columbia'
}

def state_abbrev_to_longform(abbrev):
    """Convert a US state abbreviation to its longform name."""
    return us_state_abbrev_to_name.get(abbrev.upper(), abbrev)

# Example usage:
# state_abbrev_to_longform('CA')  # Returns 'California'
```

```python
all_data_mass['state_long'] = all_data_mass['state'].apply(state_abbrev_to_longform)
all_data_mass
```

```python
all_data_mass["date_pure"]=all_data_mass["date"].astype(str).str.slice(0,10)
all_data_mass["date_pure"]
```

```python
all_data_mass.sort_values(by="date_pure", ascending=False, inplace=True)
all_data_mass
```

```python
all_data_mass["year"]=pd.to_datetime(all_data_mass["date_pure"]).dt.year
all_data_mass
```

```python
all_data_mass.year.unique()
```

```python
all_data_mass_grouped=all_data_mass.groupby(["year"])[["killed","wounded"]].sum()
```

```python
all_data_mass_grouped["victims"]=all_data_mass_grouped["killed"]+all_data_mass_grouped["wounded"]
all_data_mass_grouped
```

```python
all_data_mass_grouped.to_csv("massshooting_victims_per_year.csv")
```

```python
files_gun=glob.glob('../data/gunviolencearchive/*.csv')
files_gun
```

```python
dfs = [pd.read_csv(f) for f in files_gun]
all_data_guns = pd.concat(dfs, ignore_index=True)
all_data_guns
```

```python
all_data_guns["date_pure"]=pd.to_datetime(all_data_guns["Incident Date"],format="%B %d, %Y").astype(str)
all_data_guns.sort_values(by="date_pure", ascending=False, inplace=True)
all_data_guns
```

```python
all_data_guns["year"]=pd.to_datetime(all_data_guns["date_pure"]).dt.year
all_data_guns_grouped=all_data_guns.groupby(["year"])[["Victims Killed","Victims Injured"]].sum()
all_data_guns_grouped["Victims"]=all_data_guns_grouped["Victims Killed"]+all_data_guns_grouped["Victims Injured"]
all_data_guns_grouped
```

```python
all_data_guns_grouped.to_csv("gunviolence_victims_per_year.csv")
```

```python
matching_data=pd.merge(all_data_mass, all_data_guns, left_on=['date_pure','state_long'], right_on=['date_pure','State'],how="inner")
matching_data
```

```python
matching_data=matching_data.drop_duplicates(subset=['date_pure','state_long',"killed","wounded","Victims Killed","Victims Injured"])
```

```python
matching_data["lower_killed"]=matching_data[["killed","Victims Killed"]].min(axis=1)
matching_data["lower_wounded"]=matching_data[["wounded","Victims Injured"]].min(axis=1)
matching_data["lower_victims"]=matching_data["lower_killed"]+matching_data["lower_wounded"]
matching_data
```

```python
matching_data_grouped=matching_data.groupby(["year_y"])[["lower_killed","lower_wounded","lower_victims"]].sum()
```

```python
political_violence_data=pd.read_csv('../data/political_violence.csv')
political_violence_data["date_pure"]=pd.to_datetime(political_violence_data["Incident Date"],format="%B %d, %Y").astype(str)
political_violence_data
```

```python
political_violence_data["year"]=pd.to_datetime(political_violence_data["date_pure"]).dt.year
political_violence_data
```

```python
political_violence_data_victims=political_violence_data.groupby(["year"])[["Victims Killed","Victims Injured"]].sum()
political_violence_data_victims["Victims Total"]=political_violence_data_victims["Victims Killed"]+political_violence_data_victims["Victims Injured"]
political_violence_data_victims
```

```python
political_violence_data_victims.to_csv("political_violence_data_victims_one_source.csv")
```

```python
political_shootings=political_shootings.drop_duplicates(subset=['date_pure', 'state_long'])
political_shootings
```
