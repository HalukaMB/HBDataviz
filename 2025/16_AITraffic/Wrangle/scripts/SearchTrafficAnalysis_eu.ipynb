```python
file_eu["cleaned_main_category"]=file_eu.main_category
```

```python
def filecleaner(file):
    file["search_traffic"]=pd.to_numeric(file["search_traffic"], errors='coerce')
    file["total_ai_referrals"]=pd.to_numeric(file["total_ai_referrals"], errors='coerce')
    df_2025=file[file["yearmonth"]>="2025-01"]
    df_2024=file[(file["yearmonth"]>="2024-01")&(file["yearmonth"]<="2024-05")]

    return {"2024":df_2024,"2025":df_2025}
```

```python
dict_of_files_eu=filecleaner(file_eu)
```

```python
def mean_calc_comparer(dict_of_files):
    mean_2025_world=dict_of_files["2025"][["site","cleaned_main_category","search_traffic","total_ai_referrals","total_monthly_search_trafic"]].groupby(["site","cleaned_main_category"]).mean().reset_index()
    mean_2024_world=dict_of_files["2024"][["site","cleaned_main_category","search_traffic","total_ai_referrals","total_monthly_search_trafic"]].groupby(["site","cleaned_main_category"]).mean().reset_index()

    df_24_25_world=pd.merge(mean_2024_world, mean_2025_world, how="outer", on=["site","cleaned_main_category"], suffixes=("_2024", "_2025"))
    df_24_25_world

    return df_24_25_world
```

```python
df_eu=mean_calc_comparer(dict_of_files_eu)
df_eu
```

```python
df_24_25_eu_categories=df_eu[["cleaned_main_category","search_traffic_2024","search_traffic_2025"]].groupby("cleaned_main_category").mean()
df_24_25_eu_categories
```

```python
df_eu["search_perc_diff"]=(df_eu["search_traffic_2025"]-df_eu["search_traffic_2024"])*100/df_eu["search_traffic_2024"]
df_eu.sort_values("search_perc_diff", ascending=False, inplace=True)
df_eu
```

```python
df_eu["ai_perc_diff"]=(df_eu["total_ai_referrals_2025"]-df_eu["total_ai_referrals_2024"])*100/df_eu["total_ai_referrals_2024"]
df_eu.sort_values("ai_perc_diff", ascending=False, inplace=True)
df_eu
```

```python
df_24_25_eu_categories_red=df_24_25_eu_categories.reset_index()
df_24_25_eu_categories_red["perc_change"]=(df_24_25_eu_categories_red["search_traffic_2025"]-df_24_25_eu_categories_red["search_traffic_2024"])*100/df_24_25_eu_categories_red["search_traffic_2024"]
df_24_25_eu_categories_red.sort_values("perc_change", ascending=False, inplace=True)
df_24_25_eu_categories_red
```
