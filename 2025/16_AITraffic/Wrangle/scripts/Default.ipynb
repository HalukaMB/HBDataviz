```python
import pandas as pd
```

```python
file_world=pd.read_excel('../data/David _ Adam PR Request 2025-06.xlsx', sheet_name="Worldwide + AI Referrals")
file_world
```

```python
def filecleaner(file):
    file["search_traffic"]=pd.to_numeric(file["search_traffic"], errors='coerce')
    file["ai_referrals"]=pd.to_numeric(file["ai_referrals"], errors='coerce')
    df_2025=file[file["yearmonth"]>="2025-01"]
    df_2024=file[(file["yearmonth"]>="2024-01")&(file["yearmonth"]<="2024-05")]

    return {"2024":df_2024,"2025":df_2025}
```

```python
dict_of_files=filecleaner(file_world)
```

```python
mean_2025_world=dict_of_files["2025"][["site","main_category","ai_referrals","search_traffic","total_monthly_search_trafic"]].groupby(["site","main_category"]).mean().reset_index()
mean_2025_world
```

```python
mean_2024_world=dict_of_files["2024"][["site","search_traffic","ai_referrals","main_category","total_monthly_search_trafic"]].groupby(["site","main_category"]).mean().reset_index()
mean_2024_world
```

```python
df_24_25_world=pd.merge(mean_2024_world,mean_2025_world,on="site",suffixes=["_2024","_2025"])
df_24_25_world
```

```python
df_24_25_world[["main_category_2025","search_traffic_2024","search_traffic_2025"]].groupby("main_category_2025").mean()
```

```python
df_24_25_world["search_perc_diff"]=(df_24_25_world["search_traffic_2025"]-df_24_25_world["search_traffic_2024"])*100/df_24_25["search_traffic_2024"]
df_24_25_world.sort_values("search_perc_diff", ascending=False, inplace=True)
df_24_25_world
```

```python
df_24_25_world["search_perc_diff"]=(df_24_25_world["search_traffic_2025"]-df_24_25_world["search_traffic_2024"])*100/df_24_25["search_traffic_2024"]
df_24_25_world.sort_values("search_perc_diff", ascending=False, inplace=True)
df_24_25_world
```

```python
df_24_25_world["ai_perc_diff"]=(df_24_25_world["ai_referrals_2025"]-df_24_25_world["ai_referrals_2024"])*100/df_24_25["ai_referrals_2024"]
df_24_25_world.sort_values("ai_perc_diff", ascending=False, inplace=True)
df_24_25_world
```
