```python
file_categories=pd.read_csv('../data/double_categories_cleaned.csv', encoding='utf-8')
file_categories.rename(columns={"main_category":"cleaned_main_category"}, inplace=True)
file_categories
```

```python
file_us=pd.merge(file_us, file_categories, how='left', on='site')
file_us['cleaned_main_category'] = file_us['cleaned_main_category'].fillna(file_us['main_category'])
file_us
```

```python
file_world=pd.merge(file_world, file_categories, how='left', on='site')
file_world['cleaned_main_category'] = file_world['cleaned_main_category'].fillna(file_world['main_category'])
file_world
```

```python
def filecleaner(file):
    file["search_traffic"]=pd.to_numeric(file["search_traffic"], errors='coerce')
    file["ai_referrals"]=pd.to_numeric(file["ai_referrals"], errors='coerce')
    df_2025=file[file["yearmonth"]>="2025-01"]
    df_2024=file[(file["yearmonth"]>="2024-01")&(file["yearmonth"]<="2024-05")]

    return {"2024":df_2024,"2025":df_2025}
```

```python
dict_of_files_world=filecleaner(file_world)
```

```python
def mean_calc_comparer(dict_of_files):
    mean_2025_world=dict_of_files["2025"][["site","cleaned_main_category","search_traffic","ai_referrals","total_monthly_search_trafic"]].groupby(["site","cleaned_main_category"]).mean().reset_index()
    mean_2024_world=dict_of_files["2024"][["site","cleaned_main_category","search_traffic","ai_referrals","total_monthly_search_trafic"]].groupby(["site","cleaned_main_category"]).mean().reset_index()

    df_24_25_world=pd.merge(mean_2024_world, mean_2025_world, how="outer", on=["site","cleaned_main_category"], suffixes=("_2024", "_2025"))
    df_24_25_world

    return df_24_25_world
```

```python
df_world=mean_calc_comparer(dict_of_files_world)
df_world
```

```python
df_24_25_world_categories=df_world[["cleaned_main_category","search_traffic_2024","search_traffic_2025"]].groupby("cleaned_main_category").mean()
df_24_25_world_categories
```

```python
df_world["search_perc_diff"]=(df_world["search_traffic_2025"]-df_world["search_traffic_2024"])*100/df_world["search_traffic_2024"]
df_world.sort_values("search_perc_diff", ascending=False, inplace=True)
df_world
```

```python
df_world["ai_perc_diff"]=(df_world["ai_referrals_2025"]-df_world["ai_referrals_2024"])*100/df_world["ai_referrals_2024"]
df_world.sort_values("ai_perc_diff", ascending=False, inplace=True)
df_world
```
