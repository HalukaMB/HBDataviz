```python
import pandas as pd
```

```python
file_us=pd.read_excel('../data/David _ Adam PR Request 2025-06.xlsx', sheet_name="US + AI Referrals")
file_us.sort_values(by=["yearmonth","site"], inplace=True)
file_us
```

```python
file_eu=pd.read_excel('../data/PR Copy of Handelsblatt Search _ AI Traffic - Jul 2025.xlsx', sheet_name="Europe")
file_eu=file_eu[file_eu["yearmonth"]!= "2025-06"]
file_eu.sort_values(by=["yearmonth","site"], inplace=True)
file_eu=file_eu.rename(columns={"total_ai_referrals":"ai_referrals","main_category":"cleaned_main_category"})
file_eu
```

```python
file_categories=pd.read_csv('../data/double_categories_cleaned.csv', encoding='utf-8')
file_categories.rename(columns={"main_category":"cleaned_main_category"}, inplace=True)
file_categories
```

```python
file_us=pd.merge(file_us, file_categories, how='left', on='site')
file_us['cleaned_main_category'] = file_us['cleaned_main_category'].fillna(file_us['main_category'])
file_us
```

```python
line_chart_sum_us=file_us.groupby(["yearmonth"])["search_traffic"].sum().reset_index()
line_chart_sum_us=line_chart_sum_us[line_chart_sum_us["yearmonth"]>="2024-05"].reset_index(drop=True)
line_chart_sum_us["perc"]= line_chart_sum_us["search_traffic"]*100/line_chart_sum_us["search_traffic"].iloc[0]
line_chart_sum_us
```

```python
line_chart_sum_eu=file_eu.groupby(["yearmonth"])["search_traffic"].sum().reset_index()
line_chart_sum_eu=line_chart_sum_eu[line_chart_sum_eu["yearmonth"]>="2024-05"].reset_index(drop=True)

line_chart_sum_eu["perc"]= line_chart_sum_eu["search_traffic"]*100/line_chart_sum_eu["search_traffic"].iloc[0]
line_chart_sum_eu
```

```python
line_chart_merged=pd.merge(line_chart_sum_us,line_chart_sum_eu,on="yearmonth", suffixes=('_us', '_eu')    )
line_chart_merged
```

```python
line_chart_merged.to_csv("linechart_merged.csv",index=False)
```

```python
def filecleaner(file):
    file["search_traffic"]=pd.to_numeric(file["search_traffic"], errors='coerce')
    file["ai_referrals"]=pd.to_numeric(file["ai_referrals"], errors='coerce')
    df_2025=file[(file["yearmonth"]>="2025-01")&(file["yearmonth"]<="2025-05")]
    df_2024=file[(file["yearmonth"]>="2024-01")&(file["yearmonth"]<="2024-05")]

    return {"2024":df_2024,"2025":df_2025}
```

```python
dict_of_files_us=filecleaner(file_us)
```

```python
dict_of_files_eu=filecleaner(file_eu)
```

```python
def mean_calc_comparer(dict_of_files):
    mean_2025=dict_of_files["2025"][["site","cleaned_main_category","search_traffic","ai_referrals","total_monthly_search_trafic"]].groupby(["site","cleaned_main_category"]).mean().reset_index()
    mean_2024=dict_of_files["2024"][["site","cleaned_main_category","search_traffic","ai_referrals","total_monthly_search_trafic"]].groupby(["site","cleaned_main_category"]).mean().reset_index()

    df_24_25=pd.merge(mean_2024, mean_2025, how="outer", on=["site","cleaned_main_category"], suffixes=("_2024", "_2025"))
    df_24_25

    return df_24_25
```

```python
df_us=mean_calc_comparer(dict_of_files_us)
df_us
```

```python
df_eu=mean_calc_comparer(dict_of_files_eu)
df_eu
```

```python
df_24_25_us_sites=df_us[["site","cleaned_main_category","search_traffic_2024","search_traffic_2025"]].groupby(["site","cleaned_main_category"]).mean()
df_24_25_us_sites
```
