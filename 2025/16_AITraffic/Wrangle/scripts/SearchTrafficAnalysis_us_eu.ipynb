```python
def filecleaner(file):
    file["search_traffic"]=pd.to_numeric(file["search_traffic"], errors='coerce')
    file["ai_referrals"]=pd.to_numeric(file["ai_referrals"], errors='coerce')
    df_2025=file[(file["yearmonth"]>="2025-01")&(file["yearmonth"]<="2025-05")]
    df_2024=file[(file["yearmonth"]>="2024-01")&(file["yearmonth"]<="2024-05")]

    return {"2024":df_2024,"2025":df_2025}
```

```python
dict_of_files_us=filecleaner(file_us)
```

```python
dict_of_files_eu=filecleaner(file_eu)
```

```python
def mean_calc_comparer(dict_of_files):
    mean_2025=dict_of_files["2025"][["site","cleaned_main_category","search_traffic","ai_referrals","total_monthly_search_trafic"]].groupby(["site","cleaned_main_category"]).mean().reset_index()
    mean_2024=dict_of_files["2024"][["site","cleaned_main_category","search_traffic","ai_referrals","total_monthly_search_trafic"]].groupby(["site","cleaned_main_category"]).mean().reset_index()

    df_24_25=pd.merge(mean_2024, mean_2025, how="outer", on=["site","cleaned_main_category"], suffixes=("_2024", "_2025"))
    df_24_25

    return df_24_25
```

```python
df_us=mean_calc_comparer(dict_of_files_us)
df_us
```

```python
df_eu=mean_calc_comparer(dict_of_files_eu)
df_eu
```
