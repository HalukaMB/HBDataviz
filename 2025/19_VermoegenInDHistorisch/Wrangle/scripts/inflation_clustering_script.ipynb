# For the Bundesbank data

```python
inflation_data=pd.read_csv('../data/61111-0001_en.csv',sep=";",skiprows=6,header=None)
inflation_data.columns = ['Year', 'Val', 'Random', '%', 'Einheit']
inflation_data=inflation_data.dropna()
latest_value=inflation_data[inflation_data["Year"]=="2024"]["Val"]
latest_value
```

```python
bb_data=pd.read_csv("../data/Vermögensperzentile_Monatsberichte_Deutsche_Bundesbank.csv",skiprows=2,decimal=",", thousands='.')
bb_data=bb_data.dropna()
bb_data
```

```python
bb_data["PHF 2010/11"]=bb_data["PHF 2010/11"].str.replace('.','')
bb_data
```

```python
alt_cols=["Bedingte Verteilung"]
for i in bb_data.columns[1:]:
    alt_cols.append(i[4:8])
alt_cols
```

```python
bb_data.columns=alt_cols
```

```python
bb_data["perc"]=bb_data["Bedingte Verteilung"].str.split(".").str[0].astype(int)
bb_data["prozent"]=100-bb_data["perc"]
bb_data
```

```python
df_long = bb_data.melt(id_vars=['prozent'], 
                  value_vars=["2010",'2014', '2017', '2021',"2023"],
                  var_name='jahr', 
                  value_name='vermoegen_nominal')
df_long["vermoegen_nominal"]=pd.to_numeric(df_long["vermoegen_nominal"])
df_long
```

```python
years_list=df_long.jahr.unique().tolist()
```

```python
main_df_bb=pd.DataFrame()

for year in years_list:
    ref_year=year
    print(ref_year)

    ref_value=inflation_data[inflation_data["Year"]==ref_year]["Val"].values[0]


    inflation_factor_to_2024=latest_value/ref_value
    print(inflation_factor_to_2024)
    inflation_factor_to_2024=inflation_factor_to_2024.values[0]
    red_year=df_long[df_long["jahr"]==ref_year]
    red_year["vermoegen_inflation"] = red_year["vermoegen_nominal"] * inflation_factor_to_2024
    main_df_bb= pd.concat([main_df_bb, red_year], ignore_index=True)
main_df_bb
```

```python
main_df_bb["vermoegen_inflation"]=main_df_bb["vermoegen_inflation"].round(0).astype(int)
main_df_bb
```

```python
outer_array_bb=[]
#vermoegen_inflation
for perc_value in main_df_bb.prozent.unique():
    if ((perc_value <=50) & (perc_value >= 0.1)):
        print(perc_value)

        inner_array=[]
        for year in main_df_bb.jahr.unique():
            sub_df=main_df_bb[(main_df_bb.prozent==perc_value)&(main_df_bb.jahr==year)]
            untere_grenze=sub_df.vermoegen_inflation.values[0].item()
            untere_grenze=round(untere_grenze, 0)
            year=int(year)
            inner_array.append([year,untere_grenze])
        
        outer_array_bb.append({"state": perc_value.item(), "output": inner_array})




outer_array_bb
```

```python
import json

with open('bigger_percentage.json', 'w') as f:
    json.dump(outer_array_bb, f, indent=2)
```

```python
main_df=pd.DataFrame()
path="../data/Gpinter_output_HFCS_AC_B0.xlsx"
sheet_names = pd.ExcelFile(path).sheet_names
sheet_names
```

```python
for i in range(1,len(sheet_names)):
    sheet_names[i]
    data=pd.read_excel(path, sheet_name=sheet_names[i],)
    year=sheet_names[i][-4:]
    print(year)
    red_data=pd.DataFrame()
    red_data["prozent"]=(1-data["p"])*100
    red_data["untere_grenze"]=pd.to_numeric(data["thr"])
    red_data["obere_grenze"]=pd.to_numeric(data["thr"].shift(-1))
    red_data["year"]=year

    ref_year=str(year)
    ref_value=inflation_data[inflation_data["Year"]==ref_year]["Val"].values[0]
    inflation_factor_to_2024=latest_value/ref_value
    inflation_factor_to_2024=inflation_factor_to_2024.values[0]

    red_data["untere_grenze"] = red_data["untere_grenze"] * inflation_factor_to_2024
    red_data["obere_grenze"] = red_data["obere_grenze"] * inflation_factor_to_2024
    main_df= pd.concat([main_df, red_data], ignore_index=True)
    red_data.to_csv("../data/HFCS/red_data"+year+".csv", index=False)
```
