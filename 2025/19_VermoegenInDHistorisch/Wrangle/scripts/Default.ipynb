```python
inflation_data=pd.read_csv('../data/61111-0001_en.csv',sep=";",skiprows=6,header=None)
inflation_data.columns = ['Year', 'Val', 'Random', '%', 'Einheit']
inflation_data=inflation_data.dropna()
latest_value=inflation_data[inflation_data["Year"]=="2024"]["Val"]
```

```python
main_df=pd.DataFrame()
```

```python
path="../data/Gpinter_output_HFCS_AC_B0.xlsx"
sheet_names = pd.ExcelFile(path).sheet_names
sheet_names
```

```python
for i in range(1,len(sheet_names)):
    sheet_names[i]
    data=pd.read_excel(path, sheet_name=sheet_names[i],)
    year=sheet_names[i][-4:]
    print(year)
    red_data=pd.DataFrame()
    red_data["prozent"]=(1-data["p"])*100
    red_data["untere_grenze"]=pd.to_numeric(data["thr"])
    red_data["obere_grenze"]=pd.to_numeric(data["thr"].shift(-1))
    red_data["year"]=year

    ref_year=str(year)
    ref_value=inflation_data[inflation_data["Year"]==ref_year]["Val"].values[0]
    inflation_factor_to_2024=latest_value/ref_value
    inflation_factor_to_2024=inflation_factor_to_2024.values[0]

    red_data["untere_grenze"] = red_data["untere_grenze"] * inflation_factor_to_2024
    red_data["obere_grenze"] = red_data["obere_grenze"] * inflation_factor_to_2024
    main_df= pd.concat([main_df, red_data], ignore_index=True)
    red_data.to_csv("../data/HFCS/red_data"+year+".csv", index=False)
```
