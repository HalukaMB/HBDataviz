```python
import pandas as pd
import glob
```

```python
files=glob.glob('../data/*.csv')
files
```

```python
data=pd.read_csv(files[0],sep=";")
```

```python
data.columns
```

```python
upper_25_comments=data.describe()["commentsCount"]["75%"]
```

```python
upper_25_reactions=data.describe()["reactionsCount"]["75%"]
```

```python
most_commented_posts=data[data.commentsCount>upper_25_comments]
```

```python
most_reacted_posts=data[data.reactionsCount>upper_25_reactions]
```

```python
most_vocal_posts=pd.concat([most_commented_posts,most_reacted_posts]).drop_duplicates()
most_vocal_posts
```

```python
most_vocal_posts
```

```python
frequency_topic=most_vocal_posts.groupby("politisches_thema").count()["activityUrl"].reset_index().sort_values("activityUrl",ascending=False)
frequency_topic
```
