```python
import pandas as pd
import glob
import geopandas as gpd
```

```python
files_Vergleichszeitraum=glob.glob('../data/Vergleichszeitraum/*')
files_Vergleichszeitraum
```

```python
files_Vergleichszeitraum2=glob.glob('../data/Vergleichszeitraum2/*')
files_Vergleichszeitraum2
```

```python
def read_shapefile(folderpath):
    shapefile=glob.glob(folderpath+"/*.shp")
    current_file=gpd.read_file(shapefile[0])
    return current_file
```

```python
base_file=read_shapefile(files_Vergleichszeitraum[0])
base_file
```

```python
aggregated_file=base_file.copy()
for file in files_Vergleichszeitraum [1:]:
    aggregated_file = gpd.GeoDataFrame(pd.concat([aggregated_file, read_shapefile(file)], ignore_index=True))
aggregated_file["CONFIDENCE"]=pd.to_numeric(aggregated_file["CONFIDENCE"], errors='coerce')
aggregated_file["CONFIDENCE"].fillna(0, inplace=True)
sureish_fires_vergleichszeitraum=aggregated_file[aggregated_file["CONFIDENCE"]>95]
sureish_fires_vergleichszeitraum
```

```python
from sklearn.cluster import DBSCAN
import numpy as np

# Project to a metric CRS (meters)
gdf_metric = sureish_fires_vergleichszeitraum.to_crs(epsg=3857)

# Extract coordinates
coords = np.array(list(zip(gdf_metric.geometry.x, gdf_metric.geometry.y)))

# DBSCAN clustering with 1km (1000 meters) epsilon
db = DBSCAN(eps=2000, min_samples=1).fit(coords)
gdf_metric['cluster'] = db.labels_

# Get representative point for each cluster (e.g., centroid)
clustered = gdf_metric.dissolve(by='cluster').centroid

# Convert back to original CRS
clustered_gdf = gpd.GeoDataFrame(geometry=clustered, crs=gdf_metric.crs).to_crs(sureish_fires_vergleichszeitraum.crs)

print(clustered_gdf)
```

```python
clustered_gdf["geometry"].to_file("vergleichszeitraum.geojson", driver="GeoJSON")
```
