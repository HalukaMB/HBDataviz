```python
def dataread(filename):
    postdata_name=filename
    predata_name=postdata_name.replace('POST','PRE')
    postdata=pd.read_csv(postdata_name,thousands=',',decimal='.')
    predata=pd.read_csv(predata_name,thousands=',',decimal='.')
    return {"pre": predata, "post": postdata}
```

```python
databundle_tiktoks=dataread('../data/POST/tiktok_postings.csv')
databundle_tiktoks
```

```python
databundle_tiktoks["post"]
```

```python
df_postings_tiktok=pd.merge(predata,postdata, on="Firstname Of Influencer",how="outer",suffixes=('_pre','_post'))
df_postings_tiktok
```

```python
df_postings_tiktok.loc[8] = df_postings_tiktok.iloc[1].fillna(0) + df_postings_tiktok.iloc[5].fillna(0)
df_postings_tiktok
```

```python

df_postings_tiktok=df_postings_tiktok.rename(columns={"Firstname Of Influencer":"Name"})
df_postings_tiktok["Name"]=df_postings_tiktok["Name"].str.replace("Alice","Alice Weidel")
df_postings_tiktok["Name"]=df_postings_tiktok["Name"].str.replace("Christian","Christian Dürr")
df_postings_tiktok["Name"]=df_postings_tiktok["Name"].str.replace("Dr.","")

df_postings_tiktok["Name"]=df_postings_tiktok["Name"].str.replace("Felix","Felix Banaszak")
df_postings_tiktok["Name"]=df_postings_tiktok["Name"].str.replace("Heidi","Heidi Reichineck")
df_postings_tiktok["Name"]=df_postings_tiktok["Name"].str.replace("Lars","Lars Klingbeil")
df_postings_tiktok["Name"]=df_postings_tiktok["Name"].str.replace("BundeskanzlerFriedrich","Friedrich Merz / Bundeskanzler")

df_postings_tiktok
```

```python
df_postings_tiktok.to_csv("tiktok_postings_merged.csv", index=False)
```

```python
likes_tiktoks=dataread('../data/POST/tiktok_likes_per_month_1.csv')
likes_tiktoks["pre"]
```

```python
likes_tiktoks["pre"].columns=likes_tiktoks["pre"].iloc[0]
likes_tiktoks["pre"]=likes_tiktoks["pre"].iloc[2:]
likes_tiktoks["pre"]
```

```python
likes_tiktoks["pre"][likes_tiktoks["pre"].columns[1:]]=likes_tiktoks["pre"][likes_tiktoks["pre"].columns[1:]].applymap(lambda x: str(x).replace(',', ''))
likes_tiktoks["pre"][likes_tiktoks["pre"].columns[1:]]=likes_tiktoks["pre"][likes_tiktoks["pre"].columns[1:]].applymap(pd.to_numeric, errors='coerce')

likes_tiktoks["pre"]
```

```python
likes_tiktoks["post"].columns=likes_tiktoks["post"].iloc[0]
likes_tiktoks["post"]=likes_tiktoks["post"].iloc[2:]
likes_tiktoks["post"]
```

```python
likes_tiktoks["post"][likes_tiktoks["post"].columns[1:]]=likes_tiktoks["post"][likes_tiktoks["post"].columns[1:]].applymap(lambda x: str(x).replace(',', ''))
likes_tiktoks["post"][likes_tiktoks["post"].columns[1:]]=likes_tiktoks["post"][likes_tiktoks["post"].columns[1:]].applymap(pd.to_numeric, errors='coerce')

likes_tiktoks["post"]
```

```python
df_pre_post_likes=pd.concat([likes_tiktoks["pre"], likes_tiktoks["post"]], ).sort_values(by=["Firstname Of Influencer"])
df_pre_post_likes
```
