```python
def dataread(filename):
    postdata_name=filename
    predata_name=postdata_name.replace('POST','PRE')
    postdata=pd.read_csv(postdata_name,thousands=',',decimal='.')
    predata=pd.read_csv(predata_name,thousands=',',decimal='.')
    return {"pre": predata, "post": postdata}
```

```python
databundle_insta=dataread('../data/POST_02/instagram_postings.csv')
databundle_insta
```

```python
databundle_insta["post"]
```
