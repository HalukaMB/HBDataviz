```python
import pandas as pd
import glob
```

```python
files=glob.glob('../data/*.xlsx')
files
```

```python
data=pd.read_excel(files[0],skiprows=2)
data
```

```python
data.rename(columns={'Unnamed: 2': 'Date'}, inplace=True)
data
```

```python
data.dropna(subset=['Date'], inplace=True)
data_clean=data.iloc[1:]
data_clean
```

```python
data_clean['Date'].iloc[-1]
```

```python
data_clean['Date'] = pd.to_datetime(data_clean['Date'])
data_clean[["WCI-SHA-LAX","WCI-SHA-NYC","WCI-RTM-NYC"]]=data_clean[["WCI-SHA-LAX","WCI-SHA-NYC","WCI-RTM-NYC"]].apply(pd.to_numeric, errors='coerce')

```

```python
data_2025_april=data_clean[data_clean['Date']>="2025-03-26"]
data_2025_april.reset_index(drop=True, inplace=True)
data_2025_april
```

```python
data_2025_april_red=data_2025_april[["WCI-SHA-LAX","WCI-SHA-NYC","WCI-RTM-NYC"]].reset_index(drop=True)
change_perc_april_2025=data_2025_april_red/data_2025_april_red.iloc[0]*100-100
change_perc_april_2025
```

```python
libration_day=pd.concat([data_2025_april["Date"], change_perc_april_2025], axis=1)
```

```python
libration_day.to_csv('../data/liberation_day.csv', index=False)
```

```python
since_trump_election=data_clean[data_clean['Date']>="2024-10-20"]
since_trump_election_red=since_trump_election[["Date","WCI-SHA-LAX","WCI-SHA-NYC","WCI-RTM-NYC"]].reset_index(drop=True)
since_trump_election_red[["WCI-SHA-LAX","WCI-SHA-NYC","WCI-RTM-NYC"]]=since_trump_election_red[["WCI-SHA-LAX","WCI-SHA-NYC","WCI-RTM-NYC"]].apply(pd.to_numeric, errors='coerce')
since_trump_election_red.dropna(inplace=True)
change_perc_trump=since_trump_election_red[["WCI-SHA-LAX","WCI-SHA-NYC","WCI-RTM-NYC"]]/since_trump_election_red[["WCI-SHA-LAX","WCI-SHA-NYC","WCI-RTM-NYC"]].iloc[0]*100-100
under_trump=pd.concat([since_trump_election_red["Date"], change_perc_trump], axis=1)
under_trump
```

```python
under_trump.to_csv('../data/under_trump.csv', index=False)
```

```python
since_covid=data_clean[data_clean['Date']>="2022-01-01"]
since_covid.reset_index(drop=True, inplace=True)
```

```python
diff_between_weeks=pd.DataFrame()
diff_between_weeks["Date"]=since_covid["Date"]
diff_between_weeks
```

```python
diff_between_weeks[["WCI-SHA-LAX","WCI-SHA-NYC","WCI-RTM-NYC"]]=(since_covid[["WCI-SHA-LAX","WCI-SHA-NYC","WCI-RTM-NYC"]]-since_covid[["WCI-SHA-LAX","WCI-SHA-NYC","WCI-RTM-NYC"]].shift(1))/since_covid[["WCI-SHA-LAX","WCI-SHA-NYC","WCI-RTM-NYC"]].shift(1)*100
diff_between_weeks
```

```python
diff_between_weeks.dropna(inplace=True)
diff_between_weeks
```
