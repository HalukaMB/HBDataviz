```python
import pandas as pd
import glob
```

```python
files=glob.glob('../data/*.xlsx')
files
```

```python
metropolen_data_poll_old=pd.read_excel('../data/Umlandserie-2024-Daten-VonPoll.xlsx',skiprows=2)
metropolen_data_poll_old.dropna(subset=["Stadt/Landkreis"], inplace=True)
metropolen_data_poll_old
```

```python
metropolen_data_poll_old_red=metropolen_data_poll_old[["Stadt/Landkreis","Medizinische Einrichtungen je 100.000 Einwohner","Kindergärten und Grundschulen je 100.000 Einwohner"]]
metropolen_data_poll_old_red
```

```python
metropolen_data_poll=pd.read_excel('../data/250715_Handelsblatt_Umland_Metropolen_FINAL.xlsx',skiprows=2)
metropolen_data_poll
```

```python
metropolen_data_poll=metropolen_data_poll.dropna(subset=["Stadt/Landkreis"])
metropolen_data_poll
```

```python
metropolen_merged=pd.merge( metropolen_data_poll,metropolen_data_poll_old_red, on="Stadt/Landkreis", how="left")
metropolen_merged
```

```python
metropolen_with_data=metropolen_merged[~metropolen_merged["Medizinische Einrichtungen je 100.000 Einwohner"].isna()]
metropolen_with_data_filler=metropolen_with_data.groupby("Metropole")[["Medizinische Einrichtungen je 100.000 Einwohner","Kindergärten und Grundschulen je 100.000 Einwohner"]].mean().reset_index()
metropolen_with_data_filler
```

```python
metropolen_no_data=metropolen_merged[metropolen_merged["Medizinische Einrichtungen je 100.000 Einwohner"].isna()]
metropolen_no_data
```

```python
metropolen_no_data_subst=pd.merge(metropolen_no_data,metropolen_with_data_filler, on="Metropole", how="left")
metropolen_no_data_subst["Medizinische Einrichtungen je 100.000 Einwohner"]=metropolen_no_data_subst["Medizinische Einrichtungen je 100.000 Einwohner_y"].copy()
metropolen_no_data_subst["Kindergärten und Grundschulen je 100.000 Einwohner"]=metropolen_no_data_subst["Kindergärten und Grundschulen je 100.000 Einwohner_y"].copy()
metropolen_no_data_subst
```

```python
metropolen_no_data_clean=metropolen_no_data_subst.drop(columns=["Medizinische Einrichtungen je 100.000 Einwohner_x","Kindergärten und Grundschulen je 100.000 Einwohner_x","Medizinische Einrichtungen je 100.000 Einwohner_y","Kindergärten und Grundschulen je 100.000 Einwohner_y"])
metropolen_no_data_clean
```

```python
metropolen_poll_cleaned=pd.concat([metropolen_with_data,metropolen_no_data_clean], ignore_index=True)
metropolen_poll_cleaned
```

```python
metropolen_poll_cleaned["Stadt/Landkreis"].tolist()
```

```python
metropolen_poll_cleaned["Stadt/Landkreis"] = metropolen_poll_cleaned["Stadt/Landkreis"].str.replace("LK", "").str.strip()
metropolen_poll_cleaned["Stadt/Landkreis"] = metropolen_poll_cleaned["Stadt/Landkreis"].str.replace("-", "-").str.strip()
metropolen_poll_cleaned
```

```python
sparda_daten=pd.read_excel('../data/SPARDA_Wohnen_in_Deutschland_2025_Handelsblatt.xlsx')
sparda_daten["Kreiskennziffer"]= sparda_daten["Kreiskennziffer"].astype(str).str.zfill(5)
sparda_daten.columns
```

```python
metropolen_poll_sparda=pd.merge(metropolen_poll_cleaned, sparda_daten, how='outer', left_on='Stadt/Landkreis', right_on='Stadt/Landkreis',indicator=True)
metropolen_poll_sparda
```

```python
# so the problem are münchen and leipzig because we removed the LK
#to get better matches but now we have cases where the kreis is matched to the city
#or vice versa for these two cities
# we clean this by matching the higher price with the ags of the kreis and the lower price with the ags of the city
#and drop these two rows per city from the dataset
```

```python
muenchen_duplicates=metropolen_poll_sparda_cleaned_red[metropolen_poll_sparda_cleaned_red["Stadt/Landkreis"]=="München"]
muenchen_duplicates
```

```python
price_kreis_ags_munich=muenchen_duplicates[(muenchen_duplicates["durchschn. Kaufpreis 1. HJ 2025"]==6974)&(muenchen_duplicates["Kreiskennziffer"]=="09162")]
price_munich_ags_kreis=muenchen_duplicates[(muenchen_duplicates["durchschn. Kaufpreis 1. HJ 2025"]==8702)&(muenchen_duplicates["Kreiskennziffer"]=="09184")]
price_munich_ags_kreis
```

```python
metropolen_poll_sparda_cleaned_red.drop([price_munich_ags_kreis.index[0], price_kreis_ags_munich.index[0]],inplace=True)
metropolen_poll_sparda_cleaned_red
```

```python
leipzig_duplicates=metropolen_poll_sparda_cleaned_red[metropolen_poll_sparda_cleaned_red["Stadt/Landkreis"]=="Leipzig"]
leipzig_duplicates
```

```python
price_kreis_ags_leipzig=leipzig_duplicates[(leipzig_duplicates["durchschn. Kaufpreis 1. HJ 2025"]==2026)&(leipzig_duplicates["Kreiskennziffer"]=="14713")]
price_leipzig_ags_kreis=leipzig_duplicates[(leipzig_duplicates["durchschn. Kaufpreis 1. HJ 2025"]==3114)&(leipzig_duplicates["Kreiskennziffer"]=="14729")]
price_kreis_ags_leipzig
```

```python
metropolen_poll_sparda_cleaned_red.drop([price_leipzig_ags_kreis.index[0], price_kreis_ags_leipzig.index[0]],inplace=True)
metropolen_poll_sparda_cleaned_red
```
