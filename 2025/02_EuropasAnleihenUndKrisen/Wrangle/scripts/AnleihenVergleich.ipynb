```python
election_dates_cols=yield_to_change.columns[yield_to_change.columns.str.contains("B")]
election_dates_cols
```

```python
re_election_dates=["Country"]
re_election_dates.extend(election_dates_cols)
re_election_dates
```

```python
df=pd.DataFrame()
for index, row in yield_to_change[re_election_dates].iterrows():
    country=(row["Country"])
    list_of_snapelections=(row.dropna().tolist()[1:])
    if len(list_of_snapelections)>0:
        for snapelection in list_of_snapelections:
            snapelection_date=(snapelection[:6]+"20"+snapelection[6:])
            snapelection_date=pd.to_datetime(snapelection_date,format='%d.%m.%Y')
            after_date = snapelection_date + pd.Timedelta(days=days_to_check)
            before_date = snapelection_date - pd.Timedelta(days=days_to_check)
            election_yield_df=yields_data[(yields_data["Name"]>before_date)&(yields_data["Name"]<after_date)]
            yield_df_country_col=(election_yield_df.columns[election_yield_df.columns.str.contains(country)])
            cols=yield_df_country_col.tolist()
            snapelection_date_str=snapelection_date.strftime('%d.%m.%Y')
            sub_df=election_yield_df[cols]
            sub_df=sub_df.rename(columns={cols[0]:country+"_"+snapelection_date_str})

            sub_df=sub_df.reset_index(drop=True)
            df=pd.concat([df,sub_df],axis=1)
```

```python
df=pd.DataFrame()
for i in range(len(last_years)):
    end_of_200_days=last_years["Name"].iloc[i]+ pd.Timedelta(days=days_to_check*2)
    if last_years["Name"].iloc[-1]>end_of_200_days:
        vals_end_of_200_days=last_years[last_years["Name"]<end_of_200_days].iloc[-1]
        diff_series=last_years.iloc[i]-vals_end_of_200_days
        df=pd.concat([df,diff_series],axis=1)
df=df.T
df
```
