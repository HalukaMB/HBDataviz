```python
import pandas as pd
import glob
```

```python
files=glob.glob('../data/DE_Market/*.csv')
files
```

```python
data=pd.read_csv(files[1],skiprows=0)
```

```python
data.columns
```

```python
data
```

```python
data["Berufsspezifische Arbeitslosenquoten"]=data["Berufsspezifische Arbeitslosenquoten"].str.replace(',','.')
data["Berufsspezifische Arbeitslosenquoten"]=pd.to_numeric(data["Berufsspezifische Arbeitslosenquoten"], errors='coerce')
data
```

```python
pivoted=pd.pivot_table(data, index=['Berufe'], columns=['Jahr'], values='Berufsspezifische Arbeitslosenquoten')
pivoted=pivoted.reset_index()
pivoted.columns=["Berufe","2021","2022","2023","2024"]
pivoted
```

```python
pivoted=pivoted.dropna()
```

```python
df_change=pivoted.copy()
for index, row in pivoted.iterrows():
    df_change.at[index, '2024'] = (row['2024'] - row['2023'])

    df_change.at[index, '2023'] = (row['2023'] - row['2022'])
df_change.drop(columns=["2022",'2021'], inplace=True)
df_change
```

```python
df_change_increase=df_change[(df_change[df_change.columns[1:]]>0).all(axis=1)]
df_change_increase
```

```python
df_change_increase[(df_change_increase["2023"]>0.3)&(df_change_increase["2024"]>0.4)]
```

```python
german_jobs_df=df_change_increase[(df_change_increase["2023"]>0.3)&(df_change_increase["2024"]>0.4)]
german_jobs_df
```

```python
german_jobs_df
german_jobs_df = german_jobs_df.applymap(lambda x: x.replace(',', ', ') if isinstance(x, str) else x)
german_jobs_df["Berufe"] = german_jobs_df["Berufe"].str.replace(r'\d+', '', regex=True)
german_jobs_df
```

```python
len(german_jobs_df)
```

```python
german_jobs_df.to_csv('german_jobs.csv', index=False)
```

```python
files=glob.glob('../data/US_Market/cpsaat25*.xlsx')
files.sort()
files
```

```python
df_general=pd.read_excel(files[0],skiprows=3)
df_general=df_general[["Occupation","Unnamed: 4"]]
df_general.columns=["Occupation",files[0][27:-5]]
for file in files[1:]:
    suffix=(file[27:-5])
    occupation_df=pd.read_excel(file,skiprows=3)
    occupation_df_red=occupation_df[["Occupation","Unnamed: 4"]]
    occupation_df_red.columns=["Occupation",suffix]

    df_general=pd.merge(df_general, occupation_df_red,on="Occupation", how="outer", suffixes=('', f'_{suffix}'))
df_general
```

```python
df_general[df_general.columns[1:]]=df_general[df_general.columns[1:]].apply(pd.to_numeric, errors='coerce')
df_general_red=df_general.dropna()
df_general_red
```

```python
df_change_us=df_general_red.copy()
for index, row in df_general_red.iterrows():
    df_change_us.at[index, '2024'] = (row['2024'] - row['2023'])

    df_change_us.at[index, '2023'] = (row['2023'] - row['2022'])
df_change_us.drop(columns=["2022",'2021'], inplace=True)
df_change_us
```

```python
df_change_increase_us=df_change_us[(df_change_us[df_change_us.columns[1:]]>0).all(axis=1)]
df_change_increase_us
```

```python
us_by_occupation=df_change_increase_us[df_change_increase_us["2024"]>0.4]
us_by_occupation
```

```python
us_by_occupation.to_csv('us_unemployment_by_occupation.csv', index=False)
```
