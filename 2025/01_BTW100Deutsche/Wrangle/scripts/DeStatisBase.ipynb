```python
import pandas as pd
import numpy as np
```

```python
#this is zensus data for all german municiaplitiesy by nationality, gender, age
zensus_data=pd.read_csv("../data/DeStatisData/1000A-3037_de_flat.csv",sep=";", dtype=str)
zensus_data
```

```python
#we rename these columns to make them more readable
zensus_data.columns
```

```python
zensus_data_reduced=zensus_data[[
       '1_variable_attribute_code', '1_variable_attribute_label',
       '2_variable_attribute_label', '3_variable_attribute_label',
       '4_variable_attribute_label', 'value', 
       ]]
zensus_data_reduced.columns=["ARS","Name","Altersgruppe","Geschlecht","Nationalität","Zahl"]
#and we filter for the overall number across genders
zensus_data_reduced=zensus_data_reduced[(zensus_data_reduced["Geschlecht"]!="Insgesamt")&(zensus_data_reduced["Altersgruppe"]!="14 bis 17 Jahre")]
zensus_data_reduced=zensus_data_reduced.drop_duplicates()
```

```python
#and take only the data for all of germany
zensus_data_germany_all=zensus_data_reduced[(zensus_data_reduced["ARS"]=="DG")&(zensus_data_reduced["Altersgruppe"]!="14 bis 17 Jahre")].drop_duplicates()
zensus_data_germany_all["Zahl"]=pd.to_numeric(zensus_data_germany_all["Zahl"])
zensus_data_germany_all
```

```python
#then additionally we do a more detailed version of all this based on municipalities
zensus_data_reduced["AGS"]=zensus_data_reduced["ARS"].str[:-7]+zensus_data_reduced["ARS"].str[-3:]
zensus_data_reduced
```

```python
# and we take in a different dataset based on municipalities that classifies them by how densely populated they are
stadt_data=pd.read_excel("../data/DeStatisData/AuszugGV3QAktuell.xlsx", skiprows=3,dtype=str, sheet_name="Onlineprodukt_Gemeinden30092024")
stadt_data=stadt_data.dropna(subset="Unnamed: 7")
stadt_data
```

```python
#so we end up having a dataset that says for each municipality how densely populated it is
stadt_data["AGS"]=stadt_data["Land"]+stadt_data["RB"]+stadt_data["Kreis"]+stadt_data["Gem"]
stadt_data=stadt_data.dropna(subset=["AGS"])
stadt_data_red=stadt_data[["AGS","Unnamed: 20"]]
stadt_data_red.columns=["AGS","Besiedelung"]
stadt_data_red
```

```python
#we merge this list with the zensus data
zensus_data_red_vars=pd.merge(zensus_data_reduced,stadt_data_red,on="AGS")
zensus_data_red_vars=zensus_data_red_vars[['AGS','Name', 'Altersgruppe', 'Geschlecht', 
        'Besiedelung', 'Zahl']]
zensus_data_red_vars["Zahl"]=pd.to_numeric(zensus_data_red_vars["Zahl"],errors="coerce")
```

```python
# and then we also classify using the AGS code whether a municipality is in the east or west of germany
# berlin is excluded later from this calculation
zensus_data_red_vars["OstWest"]="Berlin"
zensus_data_red_vars["OstWest"]=np.where(zensus_data_red_vars["AGS"].str[:2].isin(["12","13","14","15","16"]),"Ost",zensus_data_red_vars["OstWest"])
zensus_data_red_vars["OstWest"]=np.where(zensus_data_red_vars["AGS"].str[:2].isin(['01', '02', '03', '04', '05', '06', '07', '08', '09', '10']),"West",zensus_data_red_vars["OstWest"])

zensus_data_red_vars.columns
```

```python
zensus_data_to_sum=zensus_data_red_vars[['Altersgruppe', 'Geschlecht', 
        'Besiedelung',"OstWest", 'Zahl']]
zensus_data_to_sum.dtypes
```

```python
#we change the column type to numeric
zensus_data_germany_all["Zahl"]=pd.to_numeric(zensus_data_germany_all["Zahl"])
sum_all=zensus_data_germany_all["Zahl"].sum()
sum_all
```

```python
#and we calculate the share of each group in the total population
summed_by_vars=zensus_data_to_sum.groupby(['Altersgruppe', 'Geschlecht', 'Besiedelung',
       'OstWest']).sum().reset_index()
summed_by_vars["Share"]=summed_by_vars["Zahl"]*100/sum_all
summed_by_vars
```

```python
#we redefine the agebands to make the subgroups bigger
summed_by_vars["Alter"]=None
summed_by_vars["Alter"]=np.where(summed_by_vars["Altersgruppe"].isin(['18 bis 24 Jahre', '25 bis 29 Jahre']),"U30",summed_by_vars["Alter"])
summed_by_vars["Alter"]=np.where(summed_by_vars["Altersgruppe"].isin(['30 bis 39 Jahre', '40 bis 49 Jahre',]),"30-49",summed_by_vars["Alter"])
summed_by_vars["Alter"]=np.where(summed_by_vars["Altersgruppe"].isin(['50 bis 59 Jahre','60 bis 64 Jahre']),"50-64",summed_by_vars["Alter"])
summed_by_vars["Alter"]=np.where(summed_by_vars["Altersgruppe"].isin(['65 bis 74 Jahre','75 Jahre und älter']),"Ü65",summed_by_vars["Alter"])
summed_by_vars.columns
```

```python
#and then group again
summed_by_vars_alt=summed_by_vars[['Geschlecht', 'Besiedelung', 'OstWest', 'Zahl', 'Share',
       'Alter']]
summed_by_vars_alt_group=summed_by_vars_alt.groupby(['Geschlecht', 'Besiedelung', 'OstWest',"Alter"]).sum().reset_index()
summed_by_vars_alt_group.sort_values(by="Share").head(20)
```

```python
#and we save this file as it is the base for our later stratification of the allbus data
summed_by_vars_alt_group.to_csv("zensus_model.csv",index=False)
```

```python
#also we write a json file in which we save the data in a different way
#most notably, we want to only enable at max two levels of filtering because otherwise the subgroups get to small
json_array=[]
inner_dict={"Geschlecht":None,"Besiedelung":None,"OstWest":None,"Alter":None,"Share":None}

for column in summed_by_vars_alt_group.columns[:4]:
    for valueofcolumn in (summed_by_vars_alt_group[column].unique()):
        #inner_dict={"Geschlecht":None,"Besiedelung":None,"OstWest":None,"Alter":None,"Share":None}
        subgroup=summed_by_vars_alt_group[ summed_by_vars_alt_group[column]==valueofcolumn]
        #inner_dict[column]=valueofcolumn
        #print(column)
        #print(valueofcolumn)
        #print(subgroup)
        for column_of_sub in subgroup.columns[:4]:
            for valueofcolumn_of_sub in (subgroup[column_of_sub].unique()):
                #print(column_of_sub)
                #print(valueofcolumn_of_sub)
                print(column,": ",valueofcolumn)
                print(column_of_sub,": ",valueofcolumn_of_sub)
                if(valueofcolumn=="Berlin" or valueofcolumn_of_sub=="Berlin"):
                    pass
                else:
                    sub_group_2=subgroup[subgroup[column_of_sub]==valueofcolumn_of_sub]
                    inner_dict[column]=valueofcolumn
                    inner_dict[column_of_sub]=valueofcolumn_of_sub
                    print(sub_group_2["Share"].sum())

                    inner_dict["Share"]=(sub_group_2["Share"].sum().item())
                    json_array.append(inner_dict)
                    inner_dict={"Geschlecht":None,"Besiedelung":None,"OstWest":None,"Alter":None,"Share":None}


#json_array
```

```python
#and we save this data to a json file
import json
with open('data_destatis.json', 'w', encoding='utf-8') as f:
    json.dump(json_array, f)
```

```python
#and also as a csv file
json_None_df=pd.DataFrame(json_array).sort_values(by="Share")
json_None_df=json_None_df.drop_duplicates()

json_None_df.to_csv("json_none.csv",index=False)
```
