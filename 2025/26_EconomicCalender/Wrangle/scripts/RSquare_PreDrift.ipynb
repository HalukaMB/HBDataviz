```python
import pandas as pd
import numpy as np
```

```python
base_data=pd.read_csv('../../Lit_StockJumps/Table4.csv')
base_data
```

```python
base_data["max_R²"] = base_data[["E-mini S&P 500 R²","10-year Treasury R²"]].max(axis=1)
base_data["max_R²"]
```

```python
base_data["significant"]=np.where(base_data["max_R²"]>0.05,True,False)
base_data
```

```python
base_data[["Announcement","E-mini S&P 500 R²","10-year Treasury R²","significant"]].to_csv("r2_significant_moves_2016.csv",index=False)
```
