```python
import pandas as pd
```

```python
data=pd.read_csv("../data/pollen.csv")
data
```

```python
reduced_data=data[["Date","Betula"]]
reduced_data
```

```python
more_than_zero=reduced_data[reduced_data["Betula"]>0]
```

```python
import altair as alt
```

```python
alt(source=more_than_zero).line(x="Date",y="Betula")
```

```python
alt.Chart(more_than_zero).mark_line().encode(
    x='Date',
    y="Betula"
)
```
