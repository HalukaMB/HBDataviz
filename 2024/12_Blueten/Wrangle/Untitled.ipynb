```python
import pandas as pd
```

```python
excelfile=pd.read_excel("data/bluehzeit.xlsx")
excelfile
```

```python
excelfile.columns[3]
```

```python
excelfile["Anfang100"]=excelfile[excelfile.columns[3]]+(excelfile["JAHR"]).astype(str)
excelfile["Ende100"]=excelfile["Unnamed: 8"]+(excelfile["JAHR"]).astype(str)
excelfile["Anfang2024"]=excelfile[excelfile.columns[3]]+"2024"
excelfile["Ende2024"]=excelfile["Unnamed: 8"]+"2024"

excelfile
```

```python
excelfile["Ende100"]=pd.to_datetime(excelfile["Ende100"],format="%d.%m.%Y")
excelfile["Anfang100"]=pd.to_datetime(excelfile["Anfang100"],format="%d.%m.%Y")
excelfile["Anfang2024"]=pd.to_datetime(excelfile["Anfang2024"],format="%d.%m.%Y")
excelfile["Ende2024"]=pd.to_datetime(excelfile["Ende2024"],format="%d.%m.%Y")

excelfile["Dauer2024"]=excelfile["Ende100"]-excelfile["Anfang100"]
excelfile
```

```python
excelfile.to_csv("bluehzeit.csv",index=False)
```

```python
excelfile["decade"]=excelfile["JAHR"].astype(str).str[:3]+"0"
decades_start=excelfile[["decade","Anfang2024"]].groupby("decade").mean()
decades_start=decades_start.reset_index()
decades_start
```

```python
decades_start.to_csv("start_of_bluethe.csv",index=False)
```

```python
excelfiledropped=excelfile.dropna(subset=["Dauer2024"])
```

```python
excelfiledropped["decade"]=excelfiledropped["JAHR"].astype(str).str[2]+"0"
excelfiledropped[["decade","Anfang2024","Ende2024","Dauer2024"]].groupby("decade").mean()
```

```python
excelfiledropped[["JAHR","Anfang2024","Ende2024","Dauer2024"]]
```

```python
excelfiledropped["Dauer2024"]=excelfiledropped["Dauer2024"].dt.days
```

```python
excelfiledropped[["JAHR","Anfang2024","Ende2024","Dauer2024"]].to_csv("stripplot.csv")
```

```python
excelfiledropped[["decade","Anfang2024","Ende2024","Dauer2024"]].groupby("decade").mean().to_csv("decadewise.csv")
```
