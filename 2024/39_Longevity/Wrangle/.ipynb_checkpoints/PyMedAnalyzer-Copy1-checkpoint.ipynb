```python
import pandas as pd
import glob
import json
```

```python
for file in glob.glob("robotics.json"):
    print(file)
    with open(file) as f:
        data = json.load(f)
        #print(data[0])
        article_df={
            "title": [],
            "pubmed_id": [],
            "firstname": [],
            "lastname": [],
            "country": [],
            "date":[]
        }
        
        for article in data:
            title=article["title"]
            date=article["publication_date"]
            pubmed_id=article["pubmed_id"]
            authors=article["authors"]
            for author in authors:
                authors_array=[]
                if"affiliation" in author:
                    if (author["affiliation"]!=None):
                        country=author["affiliation"].split(",")[-1].strip().split(".")[0]
                else:
                    country=None
                lastname=author["lastname"]
                firstname=author["firstname"]
                article_df["title"].append(title)
                article_df["pubmed_id"].append(pubmed_id)
        
                article_df["firstname"].append(firstname)
                article_df["lastname"].append(lastname)
                article_df["country"].append(country)
                article_df["date"].append(date)

        
        
        authors_df=pd.DataFrame(article_df)
        authors_df.to_csv(file[:-5]+".csv")

```

```python
df=pd.DataFrame()

for file in glob.glob("robotics.csv"):
    print(file)
    data=pd.read_csv(file)
    data["searchterm"]=file[:-4]
    df=pd.concat([df,data])
```

```python
dict_of_us_years={}
for year in range(2000,2025,1):
    filtered_df = us_mentioned[us_mentioned['date-format'].dt.year == year]
    
    dict_of_us_years[year]=[len(filtered_df["pubmed_id"].unique())]
    len(filtered_df)
```

```python
us_years=pd.DataFrame(dict_of_us_years).T
us_years.columns=["us"]
```

```python
dict_of_ch_years={}
for year in range(2000,2025,1):
    filtered_df = ch_mentioned[ch_mentioned['date-format'].dt.year == year]
    
    dict_of_ch_years[year]=[len(filtered_df["pubmed_id"].unique())]
    len(filtered_df)
```

```python
ch_years=pd.DataFrame(dict_of_ch_years).T
ch_years.columns=["ch"]
ch_years
```

```python
dict_of_de_years={}
for year in range(2000,2025,1):
    filtered_df = de_mentioned[de_mentioned['date-format'].dt.year == year]
    
    dict_of_de_years[year]=[len(filtered_df["pubmed_id"].unique())]
    len(filtered_df)
```

```python
de_years=pd.DataFrame(dict_of_de_years).T
de_years.columns=["de"]
de_years
```

```python
dict_of_overall_years={}
for year in range(2000,2025,1):
    filtered_df = df[df['date-format'].dt.year == year]
    dict_of_overall_years[year]=[len(filtered_df["pubmed_id"].unique())]
    len(filtered_df)
```

```python
overall_years=pd.DataFrame(dict_of_overall_years).T
overall_years.columns=["overall"]
```

```python
combined_df=pd.concat([us_years,ch_years,de_years,overall_years],axis=1)
combined_df
```

```python
combined_df["ch_share"]=combined_df["ch"]*100/combined_df["overall"]
combined_df["de_share"]=combined_df["de"]*100/combined_df["overall"]
combined_df["us_share"]=combined_df["us"]*100/combined_df["overall"]
```
