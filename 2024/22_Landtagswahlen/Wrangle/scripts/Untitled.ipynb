```python
import pandas as pd
from bs4 import BeautifulSoup  
import numpy as np
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.wahlrecht.de/umfragen/landtage/sachsen.htm'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

table = soup.find('table', {'class': 'wilko'})
rows = table.find_all('tr')

columnnames=[]
array=[]
for row in rows:
    columns = row.find_all('th')
    if columns:
        columnnames=[column.text.strip() for column in columns]
    cells = row.find_all('td')
    if cells:
        array.append([cell.text.strip() for cell in cells])
df_sachsen=pd.DataFrame(columns=columnnames,data=array)
df_sachsen
```

```python
splitted_sonstige=df_sachsen["Sonstige"].astype(str).str.split("%")
array_of_keys=[]
for row in range(len(splitted_sonstige)):
    for char in splitted_sonstige.iloc[row]:
        party_sonstige_list_results=char.split()
        if (len(party_sonstige_list_results)>1):
            array_of_keys.append(char.split()[0])
            
array_of_keys=list(set(array_of_keys))
splitted_sonstige=df_sachsen["Sonstige"].astype(str).str.split("%")
array_sonstige=[]
default_dict={}
for keyname in array_of_keys:
    default_dict[keyname]=0
default_dict
for row in range(len(splitted_sonstige)):
    row_dict=default_dict.copy()
    for char in splitted_sonstige.iloc[row]:
        party_sonstige_list_results=char.split()

        if (len(party_sonstige_list_results)>1):
            row_dict[party_sonstige_list_results[0]]=party_sonstige_list_results[1]
    array_sonstige.append(row_dict)
            
sonstige_df=pd.DataFrame(array_sonstige)
df_sachsen_sonstige=pd.concat([df_sachsen,sonstige_df],axis=1)
df_sachsen_polls=df_sachsen_sonstige[~df_sachsen_sonstige.Institut.str.contains("Landtag")]
df_sachsen_elections=df_sachsen_sonstige[df_sachsen_sonstige.Institut.str.contains("Landtag")]

df_sachsen_polls
```

```python
df_sachsen_polls[df_sachsen_polls.columns[4:]]=df_sachsen_polls[df_sachsen_polls.columns[4:]].replace({'%': ''}, regex=True)
df_sachsen_polls[df_sachsen_polls.columns[4:]]=df_sachsen_polls[df_sachsen_polls.columns[4:]].replace({',': '.'}, regex=True)
df_sachsen_polls=df_sachsen_polls.drop(["Sonstige"],axis=1)
df_sachsen_polls
```
