```python
import geopandas as gpd
from shapely import wkt
import json
from shapely.geometry import shape
```

```python
import geopandas as gpd

# Provide the path to your shapefile
shapefile_path = '../data/geoBoundaries-RUS-ADM2-all/geoBoundaries-RUS-ADM2.shp'

# Use the read_file() function to import the shapefile
gdf = gpd.read_file(shapefile_path)
gdf
```

```python
gdf_exploded=gdf.explode(index_parts=False)
gdf_exploded=gdf_exploded.sort_values(by="shapeName")
gdf_exploded
```

```python
for i in gdf_exploded.shapeName.unique():
    print(i)
```

```python
name_municipality="Samara"
```

```python
reduced_df=gdf_exploded.loc[gdf.shapeName==name_municipality]
reduced_df
```

```python
reduced_df.to_wkt
```

```python
reduced_df.to_file(name_municipality+'.geojson')
```

```python
f = open(name_municipality+'.geojson',) 
with open(name_municipality+".txt", "w") as fout:
    features = json.load(f)["features"]
    for feature in features:
        geo = shape(feature["geometry"])
        # format geometry coordinates as WKT
        wkt = geo.wkt
        fout.write(wkt + "\n")
```
