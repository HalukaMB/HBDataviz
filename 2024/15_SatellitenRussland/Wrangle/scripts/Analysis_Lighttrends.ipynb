```python
import pandas as pd
import glob
```

```python
list_of_files=glob.glob("../data/Russian_Cities_Light/*.csv")
list_of_files
```

```python
df=pd.DataFrame()
for i in range(len(list_of_files)):
    name=list_of_files[i][29:-10].capitalize()
    data=pd.read_csv(list_of_files[i])
    data["date"]=data["rasterColumn"].astype(str).str[-6:-2]+"-"+data["rasterColumn"].astype(str).str[-2:]+"-01"
    data["date"]=pd.to_datetime(data["date"])
    data=data[data["date"]<"2024-01-01"]
    last_year=data.iloc[-12:]
    year_before=data.iloc[-24:-12]
    two_years_before=data.iloc[-36:-24]
    three_years_before=data.iloc[-48:-36]

    last_year["median"]=last_year["mean"].median()
    year_before["median"]=year_before["mean"].median()
    two_years_before["median"]=two_years_before["mean"].median()
    three_years_before["median"]=three_years_before["mean"].median()

    
    four_year_period=pd.concat([last_year.iloc[-1:],year_before.iloc[-1:],two_years_before.iloc[-1:],three_years_before.iloc[-1:]])
    four_year_period[name]=four_year_period["median"]
    two_two_comp=four_year_period[["date",name]]
    if (i==0):
        df=two_two_comp
    else:
        df=pd.merge(df,two_two_comp,on="date")
df
```

```python
df.to_csv("df.csv",index=False)
```
