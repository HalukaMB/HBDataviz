```python
import pandas as pd
import re
import nltk
import numpy as np
```

```python
from nltk.corpus import stopwords
german_stop_words = stopwords.words('german')
german_stop_words = [word.capitalize() for word in german_stop_words]
```

```python
digital_affin=pd.read_excel("../data/Handelsblatt_Fragebogen_email ZG digital-affin.xlsx", skiprows=2)
print_affin=pd.read_excel("../data/Handelsblatt_Fragebogen_email ZG print-affin.xlsx", skiprows=2)
```

```python
def freitext_cleaner(data):
    columns_to_use=data.columns[data.columns.str.contains("Ihre Antwort")]
    freitext_df=data[columns_to_use].dropna(thresh=1)
    freitext_df.columns=["Außenpolitik","Innenpolitik","Trendthemen"]
    return freitext_df
```

```python
freitext_digital=freitext_cleaner(digital_affin)
freitext_print=freitext_cleaner(print_affin)
```

```python
def wordsFrequ(df,columname):


    freitext_list=df[columname].dropna().tolist()
    freitext_list_allwords=[]
    for entry in freitext_list:
        entry=(entry.split())
        for word in entry:
            clean_text = re.sub(r'[^a-zA-Z0-9]', '', word)
            if(clean_text.istitle()):
                if(clean_text not in german_stop_words):
                    freitext_list_allwords.append(clean_text)
    df=pd.DataFrame({'Words':freitext_list_allwords}) 
    Newdf=pd.DataFrame(df['Words'].value_counts()).reset_index()
    Newdf["Themengebiet"]=columname
    Newdf=Newdf[Newdf["count"]>2]
    return(Newdf)
```

```python
def summary_df(df):
    columnnames=df.columns.tolist()
    df_concat=pd.DataFrame()
    for columnname in columnnames:
        new_df=wordsFrequ(df,columnname)
        df_concat=pd.concat([df_concat,new_df])
    return df_concat
```

```python
print_themen=summary_df(freitext_print)
digital_themen=summary_df(freitext_digital)
print_themen.to_csv("print_themen.csv",index=False)
digital_themen.to_csv("digital_themen.csv",index=False)
```

```python
def namefile(df):
    name_df=df[df.columns[df.columns.str.contains("Politiker")]]
    Newdf=pd.DataFrame(name_df.value_counts()).reset_index()
    Newdf=Newdf[Newdf["count"]>2]
    return Newdf
```
