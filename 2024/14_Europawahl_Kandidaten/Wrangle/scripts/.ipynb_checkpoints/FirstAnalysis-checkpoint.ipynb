```python
data_europa=pd.read_csv("../data/ew2024_bewerb/ew2024_bewerb_uft8.csv",sep=";",skiprows=8)
data_europa
```

```python
EU_born_years=data_europa[["Geburtsjahr","Gruppenname"]].groupby("Gruppenname").mean().reset_index()
EU_born_years
```

```python
EU_born_years.sort_values(by="Geburtsjahr")
```

```python
data_bund=pd.read_csv("../data/btw2021_bewerb/btw2021_bewerb_utf8.csv",sep=";",skiprows=8)
data_bund
```

```python
Bund_born_years=data_bund[["Geburtsjahr","Gruppenname"]].groupby("Gruppenname").mean().reset_index()
Bund_born_years
```

```python
pd.merge(Bund_born_years,EU_born_years,on="Gruppenname",suffixes=["_Bund","_EU"])
```
