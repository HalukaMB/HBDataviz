```python
#https://de.investing.com/indices/germany-30-historical-data
dax_base_data=pd.read_csv("../data/DAX Historische Daten.csv",dtype=str)
dax_base_data["Datum"]=pd.to_datetime(dax_base_data["Datum"],format='%d.%m.%Y')
dax_base_data
```

```python
dax_base_current_data=pd.read_csv("../data/DAX Historische Daten (2).csv",dtype=str)
dax_base_current_data["Datum"]=pd.to_datetime(dax_base_current_data["Datum"],format='%d.%m.%Y')
dax_base_current_data=dax_base_current_data.sort_values("Datum")
dax_base_current_data
```

```python
merged_historic=pd.concat([dax_base_data,dax_base_current_data],axis=0,ignore_index=True)
merged_historic
```

```python
merged_historic=merged_historic.drop_duplicates(subset=["Datum"],keep="last")
merged_historic
```

```python
merged_historic["+/- %"]=merged_historic["+/- %"].str.replace(",",".").str.replace("%","")
merged_historic["+/- %"]=pd.to_numeric(merged_historic["+/- %"])
#Zuletzt vom Vortag vs. Zuletzt jetzt#
merged_historic["+/- %"]
```

```python
merged_historic["cleaned_last"]=merged_historic["Zuletzt"].str.replace(".","").str.replace(",",".")
merged_historic["cleaned_last"]=pd.to_numeric(merged_historic["cleaned_last"])
merged_historic
```

```python
merged_historic["3_day_change"]=(merged_historic["cleaned_last"]-merged_historic["cleaned_last"].shift(3))/merged_historic["cleaned_last"].shift(3)
merged_historic
```

```python
dip_periods=merged_historic[merged_historic["3_day_change"]<-0.06]
dip_periods
```

```python
no_covid=dip_periods[dip_periods["Datum"].astype(str).str[:4]!="2020"]
no_covid
```

```python
end_trend_day=no_covid["Datum"]
len(end_trend_day)
```

```python
dax40_list=df_translate["name"].tolist()
dax40_list
```

```python
overall_df=pd.DataFrame()
for name in dax40_list:
    shortener_name=df_translate[df_translate["name"]==name].shortener.values[0]
    singledf=pd.read_csv("../historic_data/csv/"+shortener_name+".csv")
    singledf["Date"]=pd.to_datetime(singledf["Date"],format='%d-%m-%Y')
    singledf["Adjusted Close"]=pd.to_numeric(singledf["Adjusted Close"])
    
    diff_list=[]
    print(name)
    
    for row in end_trend_day:
        try:
            current_val=float(singledf[singledf["Date"]==row]["Close"].iloc[0])
            previous_index=singledf[singledf["Date"]==row]["Close"].index-3
            previous_val=float(singledf[singledf.index==previous_index[0]]["Close"].iloc[0])
            diff_list.append((current_val-previous_val)/current_val)
        except:
             diff_list.append(np.nan)
    share_df=pd.DataFrame({name:diff_list})
    overall_df=pd.concat([overall_df,share_df],axis=1)
overall_df
```
