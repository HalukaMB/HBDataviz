```python
import pandas as pd
```

```python
elections2020=pd.read_excel("../data/federalelections2020.xlsx", sheet_name="16. Appendix 1A", skiprows=7, decimal=".")
elections2020=elections2020[["Unnamed: 0","(D)","(R)"]]
elections2020=elections2020.rename(columns={"Unnamed: 0":"State"})
elections2020
```

```python
primaries2024=pd.read_csv("../data/primary_runoffs_2024.csv", sep=";", decimal=",")
primaries2024
```

```python
primaries_elections=pd.merge(primaries2024,elections2020,left_on="States",right_on="State", how="outer")
primaries_elections
```

```python
primaries_elections_dropped=primaries_elections.dropna(subset=["States","State"],how="all")
```

```python
democratic_voteshare=primaries_elections[primaries_elections["Haley"]>primaries_elections["Trump"]]["(D)"]
democratic_voteshare
```

```python
primaries_elections_dropped
```

```python
primaries_elections_dropped_all=primaries_elections_dropped.dropna(subset=["States","State"],how="any").reset_index(drop=True)
primaries_elections_dropped_all
```

```python
primaries_done=primaries_elections_dropped_all.dropna(subset=["Haley"])
primaries_done
```

```python
df_to_predict=primaries_elections_dropped_all[primaries_elections_dropped_all["Haley"].isna()]
df_to_predict
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
X = primaries_done[["(R)"]]  # Input features
y = primaries_done["Haley"]       # Target values

# Create a linear regression model
model = LinearRegression()
model.fit(X, y)
X_test = df_to_predict[["(R)"]]       # New input features for prediction
y_pred = model.predict(X_test)
y_dataset = model.predict(X)

y_pred
```

```python
y_dataset
```

```python
y_true = primaries_done["Haley"]
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_true, y_dataset)
```
