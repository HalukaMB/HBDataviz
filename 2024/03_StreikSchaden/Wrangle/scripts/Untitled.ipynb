```python
import glob
```

```python
files=glob.glob("../data/area-1hour-vessel-movements-report/*")
for file in files:
    reducedfilename=(file[43:-34])
    data=pd.read_csv(file)
    data=data.drop_duplicates(subset=["IMO"])
    data["AISTYPE_Class"]=data["AISTYPE"].astype(str).str[:1]
    data["byday"]=data["DATE TIME (UTC)"].astype(str).str[:10]
    data.to_csv("../data/NoDuplicates/"+reducedfilename+".csv")
    data.head(40).to_csv("../data/NoDuplicates/"+reducedfilename+"_head.csv")
```
