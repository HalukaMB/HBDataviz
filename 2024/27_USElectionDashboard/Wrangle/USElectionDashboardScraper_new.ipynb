```python
import pandas as pd
import requests
import io
import numpy as np
import os
import datawrapper
import pygsheets
from datetime import datetime, timedelta
import json
```

```python
natesilver_url="https://static.dwcdn.net/data/fgLCD.csv"
```

```python
response = requests.get(natesilver_url)

file_object = io.StringIO(response.content.decode('utf-8'))
nate_silver_poll_data=pd.read_csv(file_object)
nate_silver_poll_data
```

```python
nate_silver_poll_data["prev_week"]=nate_silver_poll_data["e_week"].astype(str).str.split("@")
nate_silver_poll_data["prev_week_party"]=nate_silver_poll_data["prev_week"].str[0].str[0]
nate_silver_poll_data["prev_week_sign"]=nate_silver_poll_data["prev_week"].str[0].str[1]
nate_silver_poll_data["prev_week_val"]=nate_silver_poll_data["prev_week"].str[0].str[3:]
nate_silver_poll_data["prev_week_val"]=nate_silver_poll_data["prev_week_val"].replace('', '0')
nate_silver_poll_data["prev_week_val"]=pd.to_numeric(nate_silver_poll_data["prev_week_val"],errors="coerce")

nate_silver_poll_data["prev_week_dem"]=nate_silver_poll_data["pct_trend_adjustedHarrisnow"]
nate_silver_poll_data["prev_week_rep"]=nate_silver_poll_data["pct_trend_adjustedTrumpnow"]
nate_silver_poll_data
```

```python
nate_silver_poll_data["prev_week_dem"]=np.where(nate_silver_poll_data["prev_week_party"] != "D", nate_silver_poll_data["prev_week_dem"], 
         (np.where(nate_silver_poll_data["prev_week_sign"]=="+", nate_silver_poll_data["prev_week_dem"]-nate_silver_poll_data["prev_week_val"], nate_silver_poll_data["prev_week_dem"]+nate_silver_poll_data["prev_week_val"])))


nate_silver_poll_data["prev_week_rep"]=np.where(nate_silver_poll_data["prev_week_party"] != "R", nate_silver_poll_data["prev_week_rep"], 
         (np.where(nate_silver_poll_data["prev_week_sign"]=="+", nate_silver_poll_data["prev_week_rep"]-nate_silver_poll_data["prev_week_val"], nate_silver_poll_data["prev_week_rep"]+nate_silver_poll_data["prev_week_val"])))
```

```python
nate_silver_poll_data["Diff_Previous_Week"]=nate_silver_poll_data["prev_week_rep"]-nate_silver_poll_data["prev_week_dem"]
nate_silver_poll_data["Diff_This_Week"]=nate_silver_poll_data["pct_trend_adjustedTrumpnow"]-nate_silver_poll_data["pct_trend_adjustedHarrisnow"]
nate_silver_poll_data_week_comp=nate_silver_poll_data[["Contest","Diff_Previous_Week","Diff_This_Week"]]
nate_silver_poll_data_week_comp
```

```python
nate_silver_poll_data_week_comp["Tendenz"]="Unklar / Wechselhaft"
nate_silver_poll_data_week_comp["Tendenz"]=np.where((nate_silver_poll_data_week_comp["Diff_This_Week"]<-5)&(nate_silver_poll_data_week_comp["Diff_This_Week"]<-5),"Eher demokratisch",nate_silver_poll_data_week_comp["Tendenz"])
nate_silver_poll_data_week_comp["Tendenz"]=np.where((nate_silver_poll_data_week_comp["Diff_This_Week"]>5)&(nate_silver_poll_data_week_comp["Diff_This_Week"]>5),"Eher republikanisch",nate_silver_poll_data_week_comp["Tendenz"])
nate_silver_poll_data_week_comp
```

```python
electors=pd.read_csv("data/Electors.csv")
electors
```

```python
poll_comp_week=pd.merge(nate_silver_poll_data_week_comp,electors,left_on="Contest",right_on="Jurisdiction")
poll_comp_week=poll_comp_week[["Contest","Diff_Previous_Week","Diff_This_Week","Tendenz","Electoral votes"]]
poll_comp_week=poll_comp_week.sort_values(by="Diff_This_Week",ascending=True)
poll_comp_week
```

```python
poll_comp_week.to_csv("poll_comparison.csv",index=False)
```

```python
poll_comp_week["Vorwoche_text"]=""
poll_comp_week["Vorwoche_text"]=np.where((poll_comp_week["Diff_Previous_Week"]<0),"Vorwoche eher demokratisch",poll_comp_week["Vorwoche_text"])
poll_comp_week["Vorwoche_text"]=np.where((poll_comp_week["Diff_Previous_Week"]>0),"Vorwoche eher republikanisch",poll_comp_week["Vorwoche_text"])
poll_comp_week["Jetzt_text"]=""
poll_comp_week["Jetzt_text"]=np.where((poll_comp_week["Diff_This_Week"]<0),"Jetzt eher demokratisch",poll_comp_week["Jetzt_text"])
poll_comp_week["Jetzt_text"]=np.where((poll_comp_week["Diff_This_Week"]>0),"Jetzt eher republikanisch",poll_comp_week["Jetzt_text"])
poll_comp_week["Diff_Previous_Week_abs"]=poll_comp_week["Diff_Previous_Week"].abs()
poll_comp_week["Diff_This_Week_abs"]=poll_comp_week["Diff_This_Week"].abs()

poll_comp_week=poll_comp_week.reset_index(drop=True)
equal_link="https://media.handelsblatt.com/handelsblatt/downloads/2024/divteam/09/arrows/equal2.png"
 
dem_down_link="https://media.handelsblatt.com/handelsblatt/downloads/2024/divteam/09/arrows/dem_down2.png"
 
dem_up_link="https://media.handelsblatt.com/handelsblatt/downloads/2024/divteam/09/arrows/dem_up2.png"
 
rep_down_link="https://media.handelsblatt.com/handelsblatt/downloads/2024/divteam/09/arrows/rep_down2.png"
 
rep_up_link="https://media.handelsblatt.com/handelsblatt/downloads/2024/divteam/09/arrows/rep_up2.png"
```

```python
counter=0
linklist=[]
for index, row in poll_comp_week.iterrows():
    counter+=1

    if(row["Diff_Previous_Week"]<0):
        if abs(row["Diff_This_Week"]-row["Diff_Previous_Week"])<0.1:
            linklist.append(equal_link)
        elif (row["Diff_This_Week"]<row["Diff_Previous_Week"]):
            linklist.append(dem_up_link)
        elif (row["Diff_This_Week"]>row["Diff_Previous_Week"]):
            linklist.append(dem_down_link)
        else:
            linklist.append(equal_link)

    if(row["Diff_Previous_Week"]>0):
        if abs(row["Diff_This_Week"]-row["Diff_Previous_Week"])<0.1:
            linklist.append(equal_link)
        elif (row["Diff_This_Week"]>row["Diff_Previous_Week"]):
            linklist.append(rep_up_link)
        elif (row["Diff_This_Week"]<row["Diff_Previous_Week"]):
            linklist.append(rep_down_link)
        else:
            linklist.append(equal_link)    
link_df=pd.DataFrame({"link":linklist})
poll_comp_week_link=pd.concat([poll_comp_week,link_df],axis=1)
poll_comp_week_link["Bundesstaat"]=poll_comp_week_link["Contest"]+" !["+poll_comp_week_link["link"].str[77:-4]+"]("+poll_comp_week_link["link"]+")"
#Minnesota ![down](https://media.handelsblatt.com/handelsblatt/downloads/2024/divteam/09/arrows/dem_down.png)
poll_comp_week_link=poll_comp_week_link[["Bundesstaat","Diff_Previous_Week_abs","Diff_This_Week_abs","Vorwoche_text","Jetzt_text"]]
poll_comp_week_link
poll_comp_week_link.to_csv("poll_comp_week_link.csv",index=False)
poll_comp_week["Electoral votes"].sum()
```

```python
yield_curve_csv="https://home.treasury.gov/resource-center/data-chart-center/interest-rates/daily-treasury-rates.csv/2024/all?field_tdr_date_value=2024&type=daily_treasury_yield_curve&page&_format=csv"
response = requests.get(yield_curve_csv)

file_object = io.StringIO(response.content.decode('utf-8'))
yield_curve_data=pd.read_csv(file_object)
no_months_columns=yield_curve_data.columns[~yield_curve_data.columns.str.contains("Mo")]
yield_curve_data_red=yield_curve_data[no_months_columns]
yield_curve_data_red=yield_curve_data_red.drop(["1 Yr","20 Yr","30 Yr"],axis=1)
yield_curve_data_red.columns=yield_curve_data_red.columns.str.replace(" Yr","")
current_week_yield=yield_curve_data_red[yield_curve_data_red.columns[1:]].iloc[:7].mean()
past_week_yield=yield_curve_data_red[yield_curve_data_red.columns[1:]].iloc[7:14].mean()

# In[19]:

yield_curve_comp=pd.concat([current_week_yield,past_week_yield],axis=1)
yield_curve_comp=yield_curve_comp.reset_index()
yield_curve_comp.columns=["Zeit","Aktuelle Woche", "Vorwoche"]
yield_curve_comp.Zeit=yield_curve_comp.Zeit.str.replace("Yr","J")
yield_curve_comp.Zeit=yield_curve_comp.Zeit.str.replace("Mo","M")

yield_curve_data_red_comp=yield_curve_comp
yield_curve_data_red_comp.to_csv("yield_curve_data_red.csv",index=False)

yield_curve_data_red_comp
```
