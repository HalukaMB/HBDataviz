```python
excel=pd.read_excel("data/GehaltsTabellenUpdated.xlsx",sheet_name="Couple")
excel.columns=excel.loc[0]

excel=excel.iloc[1:-3]
excel=excel.dropna(axis=1)
```

```python
excel.columns
```

```python
columns_new=['Einkommen']
for column in excel.columns[1:]:
    columns_new.append(int(str(column).strip()))
excel.columns=columns_new
excel
```

```python
transposed_couple=excel.T

transposed_couple.columns=transposed_couple.iloc[0].tolist()
transposed_couple=transposed_couple.iloc[1:]
transposed_couple=transposed_couple.rename(columns={"index":"Einkommen"})
transposed_couple[[6000,8000,9000]]
transposed_couple
```

```python
excel.index=excel["Einkommen"]
excel
```

```python
excel_restruct=pd.concat([excel,transposed_couple[[6000,8000,9000]]],axis=1)
excel_restruct=excel_restruct[[ 0, 1000, 2000, 3000, 4000, 5000, 6000,7000,  8000,
       9000,10000 ]]
excel_restruct=excel_restruct.iloc[:-1]
excel_restruct
```

```python
df_restruct=excel_restruct[[0,
 1000,
 2000,
 3000,
 4000,
 5000,
6000,
 7000,
 8000,
 9000, 
 10000]]
df_restruct
```

```python
df_restruct.columns=df_restruct.columns.astype(int)
df_restruct.index=df_restruct.index.astype(int)

df_restruct=df_restruct.apply(pd.to_numeric)
df_restruct
```

```python
df_restruct_interpolated=df_restruct.interpolate(method='polynomial', order=2, limit_direction='forward', axis=1)
df_restruct_interpolated=df_restruct_interpolated.round()
df_restruct_interpolated
```

```python
list_of_2nd_cols=df_restruct.columns
list_of_2nd_cols
```

```python
restructured_df=pd.concat([excel["Einkommen"],df_restruct],axis=1)
restructured_df
```

```python
math.isnan(restructured_df[6000].loc[6000])
```

```python
df_restruct_interpolated[6000].loc[1000]
```

```python
for index, row in restructured_df.iterrows():
    for column in list_of_2nd_cols:
        inner_dict={"status":"VerheiratetMitKind"}
        inner_dict["Gehalt1"]=int(row["Einkommen"])
        inner_dict["Gehalt2"]=column
        print(column)
        print(row[column])
        if math.isnan(row[column]):
            inner_dict["taxChange"]=df_restruct_interpolated[column].loc[index]
            inner_dict["extrapolated"]=True
            print("extrapolate")
        else:    
            inner_dict["taxChange"]=int(row[column])
            inner_dict["extrapolated"]=False



        prototype_json.append(inner_dict)
prototype_json
```

```python
with open('data_2.json', 'w') as f:
    json.dump(prototype_json, f)
```
