```python
excel=pd.read_excel("data/GehaltsTabellen.xlsx",sheet_name="VerheiratetZweiKinder")
excel=excel.iloc[:-1]
excel
```

```python
excel["6000"]=np.nan
excel["8000"]=np.nan
excel["9000"]=np.nan
excel
```

```python
excel.columns
```

```python
columns_new=['Einkommen']
for column in excel.columns[1:]:
    columns_new.append(int(str(column).strip()))
excel.columns=columns_new
excel
```

```python
excel_restruct=excel[['Einkommen',
 0,
 1000,
 2000,
 3000,
 4000,
 5000,
6000,
 7000,
 8000,
 9000, 
 10000]]
excel_restruct
```

```python
df_restruct=excel_restruct[[0,
 1000,
 2000,
 3000,
 4000,
 5000,
6000,
 7000,
 8000,
 9000, 
 10000]]
df_restruct
```

```python
df_restruct.columns=df_restruct.columns.astype(int)
df_restruct
```

```python
df_restruct=df_restruct.interpolate(method='polynomial', order=2, limit_direction='forward', axis=1)
df_restruct=df_restruct.round()
df_restruct
```

```python
list_of_2nd_cols=df_restruct.columns
list_of_2nd_cols
```

```python
restructured_df=pd.concat([excel_restruct["Einkommen"],df_restruct],axis=1)
```

```python
for index, row in restructured_df.iterrows():
    for column in list_of_2nd_cols:
        inner_dict={"status":"VerheiratetMitKind"}
        inner_dict["Gehalt1"]=int(row["Einkommen"])
        inner_dict["Gehalt2"]=column
        print(column)
        print(row[column])

        inner_dict["taxChange"]=int(row[column])
        prototype_json.append(inner_dict)
prototype_json
```

```python
with open('data.json', 'w') as f:
    json.dump(prototype_json, f)
```
