```python
excel=pd.read_excel("data/GehaltsTabellen.xlsx",sheet_name="VerheiratetZweiKinder")
excel
```

```python
transposed_couple=excel.T

transposed_couple.columns=transposed_couple.iloc[0].tolist()
transposed_couple=transposed_couple.iloc[1:]
transposed_couple=transposed_couple.rename(columns={"index":"Einkommen"})
transposed_couple
```

```python
transposed_couple[[6000,8000,9000]]
```

```python
excel.index=excel["Einkommen"]
excel
```

```python
excel_restruct=pd.concat([excel,transposed_couple[[6000,8000,9000]]],axis=1)
excel_restruct=excel_restruct[[ 0, 1000, 2000, 3000, 4000, 5000, 6000,7000,  8000,
       9000,10000 ]]
excel_restruct=excel_restruct.iloc[:-2]
excel_restruct
```

```python
excel_restruct.columns=excel_restruct.columns.astype(int)
excel_restruct.index=excel_restruct.index.astype(int)

excel_restruct_interpolated=excel_restruct.interpolate(method='polynomial', order=2, limit_direction='forward', axis=1)
excel_restruct_interpolated
```

```python
excel_restruct_interpolated["Einkommen"]=excel_restruct_interpolated.index
excel_restruct_interpolated
```

```python
list_of_2nd_cols=excel_restruct_interpolated.columns.tolist()[:-1]
list_of_2nd_cols
```

```python
for index, row in excel_restruct_interpolated.iterrows():
    for column in list_of_2nd_cols:
        inner_dict={"status":"VerheiratetMitKind"}
        inner_dict["Gehalt1"]=int(row["Einkommen"])
        inner_dict["Gehalt2"]=column
        print(column)
        print(row[column])

        inner_dict["taxChange"]=int(row[column])
        prototype_json.append(inner_dict)
prototype_json
```

```python
with open('data.json', 'w') as f:
    json.dump(prototype_json, f)
```
