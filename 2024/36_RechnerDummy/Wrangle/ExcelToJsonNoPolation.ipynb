```python
excel=pd.read_excel("data/Entlastung 2025 SteuFo Werte HB 15.01.2025.xlsx",sheet_name="Splitting kein Kind")
excel.columns=excel.loc[1]

excel=excel.iloc[2:]
column_numbers = [x for x in range(excel.shape[1])]  # list of columns' integer indices

column_numbers .remove(0) #removing column integer index 0
excel=excel.iloc[:, column_numbers] #return all columns except the 0th column
excel=excel.rename(columns={excel.columns[0]:"Einkommen"})
col_2nd_list=excel.columns[1:].tolist()
```

```python
for index, row in excel.iterrows():
    for column in col_2nd_list:
        inner_dict={"status":"Verheiratet"}
        inner_dict["Gehalt1"]=int(row["Einkommen"])
        inner_dict["Gehalt2"]=int(column)
        #print(row[column])
        inner_dict["taxChange"]=inner_dict["taxChange"]=int(row[column])
        inner_dict["extrapolated"]=False




        prototype_json.append(inner_dict)
#prototype_json
```
