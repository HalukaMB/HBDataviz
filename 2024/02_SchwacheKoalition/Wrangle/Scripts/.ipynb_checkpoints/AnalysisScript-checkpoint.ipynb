```python
import pandas as pd
import numpy as np
```

We load the data with the data for all polls since 1991

```python
rawdata=pd.read_excel("../RawData/1_Projektion_1_2.xlsx", skiprows=7)
rawdata
```

Clean the data by dropping NaNs and renaming columns

```python
rawdata_withoutna=rawdata.dropna(subset=["CDU/CSU"])
rawdata_withoutna=rawdata_withoutna.reset_index()
rawdata_withoutna=rawdata_withoutna.rename(columns={"CDU/CSU":"CDU"})
rawdata_withoutna_columns_renamed=rawdata_withoutna.rename(columns={"Unnamed: 1":"Datum"})
rawdata_withoutna_columns_renamed.columns
basedata_cleaned=rawdata_withoutna_columns_renamed.drop(columns=["Unnamed: 0"])
basedata_cleaned
```

And then we change the date column into datetime and save the file away

```python
basedata_cleaned["Datum"]=pd.to_datetime(basedata_cleaned["Datum"])
basedata_cleaned
```

```python
basedata_cleaned.iloc[-1].to_csv("letzte_umfrage.csv")
```

We also create a separate copy only with the last poll results

```python
basedata_cleaned[basedata_cleaned.columns[2:]].iloc[-1].to_csv("letzte_umfrage.csv")
```

Then we create an extra dataframe that holds the start and end date of coalitions that reigned Germany between 1990 and now

```python
koalitionen=pd.DataFrame([[["CDU","FDP"],"1982-10-04","1998-10-27"],
 [["SPD","Grüne"],"1998-10-27","2002-10-22"],
[["SPD","Grüne"],"2002-10-22","2005-11-22"],
[["CDU","SPD"],"2005-11-22","2009-10-28"],
[["CDU","FDP"],"2009-10-28","2013-12-17"],
[["CDU","SPD"],"2013-12-17","2021-12-08"],
[["SPD","Grüne","FDP"],"2021-12-08","2025-12-08"]])
koalitionen
```

```python
koalitionen.columns=["Koalitionen","Start","Ende"]
koalitionen.Start=pd.to_datetime(koalitionen.Start)
koalitionen.Ende=pd.to_datetime(koalitionen.Ende)
koalitionen
```

And then we run a for loop to assemble our own dataframe that hold for each date only the percentage points of the coalition that was in power

```python
summation_df=pd.DataFrame()
#so we iterate through the number of coalitions held in the koalitionen dataframe
for legislaturcount in range(len(koalitionen)):
    # get the row that hold the info for the involved parties and the start and end date
    selected_legislatur=koalitionen.iloc[legislaturcount]
    #then we filter the basedataset for the time
    legislatur_df=basedata_cleaned[(selected_legislatur["Start"]<basedata_cleaned["Datum"])&(selected_legislatur["Ende"]>basedata_cleaned["Datum"])]
    #then we iterate through all columns of this time-wise reduced dataset 
    #and if the column has the name of a reigning party we keep it in a list
    koalitionsparteien = [ ele for ele in (legislatur_df.columns) if ele in selected_legislatur["Koalitionen"]]
    #we add this list to one with the name "Datum"
    df_list=["Datum"]
    df_list.extend(koalitionsparteien)
    #And hence we can filter the time-wise reduced dataset for the reigning parties
    legislatur_df_reduced=legislatur_df[df_list]
    #we retrtieve the year of the start of the coalition using the datum column
    datenyear=str(legislatur_df_reduced["Datum"].iloc[0])
    #and write this reduced dataframe to a csv
    legislatur_df_reduced.to_csv(datenyear[:4]+".csv", index=False)
    #also we add this through a general summation dataframe that holds the data for all coalitions since 1991
    summation_df=pd.concat([summation_df,legislatur_df_reduced])
```

```python
summation_df
```

As only the percentage points of the reigning parties is held, we can sum per row
to see what was the share of votes for the entire coalition

```python
summation_df["Koalition"]=summation_df[summation_df.columns[1:]].sum(axis=1)
summation_df
```

We resort the order of columns and then sort by the value for the coalitions
and we see that 6 out of the 10 worst results are from the current party.



```python
summation_df=summation_df[["Datum","CDU","SPD","FDP","Grüne","Koalition"]]
summation_df=summation_df.sort_values(by="Koalition")
summation_df.head(13)
```

We save the top results and the overall dataframe in CSV files

```python
summation_df.sort_values(by="Koalition").head(10).to_csv("koalition.csv", index=False)
```

```python
summation_df.to_csv("koalition_all.csv", index=False)
```

Next we look at the dataframe that tracks the problems that were perceived as most important.
Again some renaming, dropping of NAs etc. for both files.

```python
probleme=pd.read_excel("../RawData/9_Probleme_1.xlsx", skiprows=7)
probleme=probleme.rename(columns={"Unnamed: 1":"Datum"})
probleme=probleme.drop("Unnamed: 0", axis=1)
probleme_withoutna=probleme.dropna(subset=["Datum"])
probleme_withoutna=probleme_withoutna.reset_index(drop=True)
probleme_withoutna
```

```python
probleme2=pd.read_excel("../RawData/10_Probleme_2.xlsx", skiprows=7)
probleme2=probleme2.drop("Unnamed: 0", axis=1)
probleme2=probleme2.rename(columns={"Unnamed: 1":"Datum"})
probleme2_withoutna=probleme2.dropna(subset=["Datum"])
probleme2_withoutna=probleme2_withoutna.reset_index(drop=True)
probleme2_withoutna
```

We merge these files based on date

```python
all_problems_df=pd.merge(probleme_withoutna,probleme2_withoutna, on="Datum")
all_problems_df
```

And then we filter rows out of these dataframes that stand for time periods when the backing
of the coalitions were the lowest.

```python
all_problems_df_2004=all_problems_df[all_problems_df["Datum"].astype(str).str.contains('|'.join(["2004","2005"]))]
all_problems_df_2004=all_problems_df_2004[all_problems_df_2004["Datum"]=="2004-08-27"]
all_problems_df_2004
```

```python
all_problems_df_2010=all_problems_df[all_problems_df["Datum"].astype(str).str.contains('|'.join(["2010"]))]
all_problems_df_2010=all_problems_df_2010[all_problems_df_2010["Datum"]=="2010-10-01"]
all_problems_df_2010
```

```python
all_problems_df_2024=all_problems_df[all_problems_df["Datum"].astype(str).str.contains('|'.join(["2024","2023"]))]
```

We create an overall dataframe for all these time periods using concat

```python
current_problems=pd.DataFrame(all_problems_df_2024.iloc[-1]).T
current_problems
previous_problems=pd.concat([all_problems_df_2004,all_problems_df_2010])
previous_problems
```

```python
problems_comparison=pd.concat([previous_problems, current_problems])
problems_comparison
```

Then we drop the columns with NAs

```python
problems_comparison=problems_comparison.dropna(axis=1)
problems_comparison
```

Now we simplify the date column and transpose the dataframe

```python
problems_comparison.Datum=problems_comparison.Datum.astype(str).str[:4]
problems_comparison
```

```python
problems_flipped=problems_comparison.T
problems_flipped
```

```python
problems_flipped.iloc[0].tolist()
```

```python
problems_flipped.columns=problems_flipped.iloc[0].tolist()
problems_flipped=problems_flipped.iloc[1:]
problems_flipped
```

Then we collect for each timeframe the three most perceived problems. And we create
an overall dataframe that includes problems that only holds the info for problems that were at least in the top 3 in 2004, 2010 or 2024 once.

```python
problemlist2004=problems_flipped.sort_values(by="2004",ascending=False).index.tolist()[:3]
problemlist2010=problems_flipped.sort_values(by="2010",ascending=False).index.tolist()[:3]
problemlist2024=problems_flipped.sort_values(by="2024",ascending=False).index.tolist()[:3]
problemlist2024
```

```python
allproblems_list=problemlist2004+problemlist2010+problemlist2024
allproblems_list=list(set(allproblems_list))
allproblems_list
```

```python
problems_flipped=problems_flipped[problems_flipped.index.isin(allproblems_list)]
```

```python
problems_flipped.sort_values(by="2010",ascending=False)
```

```python
problems_flipped.sort_values(by="2024",ascending=False)
```

Some renaming of certain problems using commas instead of dashes

```python
problems_flipped.index=problems_flipped.index.str.replace("/",", ")
```

```python
problems_flipped_2004=problems_flipped.sort_values(by="2004",ascending=False).reset_index()
problems_flipped_2004
```

And writing out the dataframe

```python
problems_flipped_2004.to_csv("problemscomparison.csv", index=False, sep=";")
```
