```python
import pandas as pd
import requests
import io
import numpy as np
```

```python
natesilver_url="https://static.dwcdn.net/data/OYvXP.csv"
```

```python
response = requests.get(natesilver_url)

file_object = io.StringIO(response.content.decode('utf-8'))
nate_silver_poll_data=pd.read_csv(file_object)
nate_silver_poll_data
```

```python
nate_silver_poll_data["prev_week"]=nate_silver_poll_data["e_week"].astype(str).str.split("@")
nate_silver_poll_data["prev_week_party"]=nate_silver_poll_data["prev_week"].str[0].str[0]
nate_silver_poll_data["prev_week_sign"]=nate_silver_poll_data["prev_week"].str[0].str[1]
nate_silver_poll_data["prev_week_val"]=nate_silver_poll_data["prev_week"].str[0].str[3:]
nate_silver_poll_data["prev_week_dem"]=nate_silver_poll_data["pct_trend_adjustedHarrisnow"]
nate_silver_poll_data["prev_week_rep"]=nate_silver_poll_data["pct_trend_adjustedTrumpnow"]
nate_silver_poll_data
```

```python
nate_silver_poll_data["prev_week_dem"]=np.where(nate_silver_poll_data["prev_week_party"] != "D", nate_silver_poll_data["prev_week_dem"], 
         (np.where(nate_silver_poll_data["prev_week_sign"]=="+", nate_silver_poll_data["prev_week_dem"]-nate_silver_poll_data["prev_week_val"].replace('', '0').astype(float), nate_silver_poll_data["prev_week_dem"]+nate_silver_poll_data["prev_week_val"].replace('', '0').astype(float))))


nate_silver_poll_data["prev_week_rep"]=np.where(nate_silver_poll_data["prev_week_party"] != "R", nate_silver_poll_data["prev_week_rep"], 
         (np.where(nate_silver_poll_data["prev_week_sign"]=="+", nate_silver_poll_data["prev_week_rep"]-nate_silver_poll_data["prev_week_val"].replace('', '0').astype(float), nate_silver_poll_data["prev_week_rep"]+nate_silver_poll_data["prev_week_val"].replace('', '0').astype(float))))
```

```python
nate_silver_poll_data["prev_month"]=nate_silver_poll_data["e_month"].astype(str).str.split("@")
nate_silver_poll_data["prev_month_party"]=nate_silver_poll_data["prev_month"].str[0].str[0]
nate_silver_poll_data["prev_month_sign"]=nate_silver_poll_data["prev_month"].str[0].str[1]
nate_silver_poll_data["prev_month_val"]=nate_silver_poll_data["prev_month"].str[0].str[3:]
nate_silver_poll_data["prev_month_dem"]=nate_silver_poll_data["pct_trend_adjustedHarrisnow"]
nate_silver_poll_data["prev_month_rep"]=nate_silver_poll_data["pct_trend_adjustedTrumpnow"]
nate_silver_poll_data
```

```python
nate_silver_poll_data["prev_month_dem"]=np.where(nate_silver_poll_data["prev_month_party"] != "D", nate_silver_poll_data["prev_month_dem"], 
         (np.where(nate_silver_poll_data["prev_month_sign"]=="+", nate_silver_poll_data["prev_month_dem"]-nate_silver_poll_data["prev_month_val"].replace('', '0').astype(float), nate_silver_poll_data["prev_month_dem"]+nate_silver_poll_data["prev_month_val"].replace('', '0').astype(float))))


nate_silver_poll_data["prev_month_rep"]=np.where(nate_silver_poll_data["prev_month_party"] != "R", nate_silver_poll_data["prev_month_rep"], 
         (np.where(nate_silver_poll_data["prev_month_sign"]=="+", nate_silver_poll_data["prev_month_rep"]-nate_silver_poll_data["prev_month_val"].replace('', '0').astype(float), nate_silver_poll_data["prev_month_rep"]+nate_silver_poll_data["prev_month_val"].replace('', '0').astype(float))))
nate_silver_poll_data
```

```python
nate_silver_poll_data["Diff_Previous_Week"]=nate_silver_poll_data["prev_week_rep"]-nate_silver_poll_data["prev_week_dem"]
nate_silver_poll_data["Diff_This_Week"]=nate_silver_poll_data["pct_trend_adjustedTrumpnow"]-nate_silver_poll_data["pct_trend_adjustedHarrisnow"]
nate_silver_poll_data_week_comp=nate_silver_poll_data[["Contest","Diff_Previous_Week","Diff_This_Week"]]
nate_silver_poll_data_week_comp
```

```python
nate_silver_poll_data_week_comp["Tendenz"]="Unklar / Wechselhaft"
nate_silver_poll_data_week_comp["Tendenz"]=np.where((nate_silver_poll_data_week_comp["Diff_This_Week"]<-2)&(nate_silver_poll_data_week_comp["Diff_This_Week"]<-2),"Eher demokratisch",nate_silver_poll_data_week_comp["Tendenz"])
nate_silver_poll_data_week_comp["Tendenz"]=np.where((nate_silver_poll_data_week_comp["Diff_This_Week"]>2)&(nate_silver_poll_data_week_comp["Diff_This_Week"]>2),"Eher republikanisch",nate_silver_poll_data_week_comp["Tendenz"])
nate_silver_poll_data_week_comp
```

```python
electors=pd.read_csv("../data/Electors.csv")
electors
```

```python
poll_comp_week=pd.merge(nate_silver_poll_data_week_comp,electors,left_on="Contest",right_on="Jurisdiction")
poll_comp_week["Contest"]=poll_comp_week["Contest"]+" ("+poll_comp_week["Electoral votes"].astype(str)+")"
poll_comp_week=poll_comp_week[["Contest","Diff_Previous_Week","Diff_This_Week","Tendenz","Electoral votes"]]
poll_comp_week=poll_comp_week.sort_values(by="Electoral votes",ascending=True)
poll_comp_week
```

```python
poll_comp_week.to_csv("poll_comparison.csv",index=False)
```

```python
poll_comp_week["Electoral votes"].sum()
```

```python
base_dem=226-10-4-13
base_rep=235-16-30
base_dem+base_rep
```

```python
dem_prev_week=poll_comp_week[poll_comp_week["Diff_Previous_Week"]<-2]["Electoral votes"].sum()
rep_prev_week=poll_comp_week[poll_comp_week["Diff_Previous_Week"]>2]["Electoral votes"].sum()
dem_this_week=poll_comp_week[poll_comp_week["Diff_This_Week"]<-2]["Electoral votes"].sum()
rep_this_week=poll_comp_week[poll_comp_week["Diff_This_Week"]>2]["Electoral votes"].sum()
rep_this_week
```

```python
toss_up_prev_week=538-base_dem-base_rep-dem_prev_week-rep_prev_week
toss_up_this_week=538-base_dem-base_rep-dem_this_week-rep_this_week
toss_up_this_week
```

```python
df_electoral=pd.DataFrame([[base_dem+dem_prev_week,toss_up_prev_week,rep_prev_week+base_rep],
 [base_dem+dem_this_week,toss_up_this_week,rep_this_week+base_rep]],index=["Vorwoche", "Diese Woche"],columns=["Eher demokratisch","Unklar/Wechselhaft","Eher republikanisch"])
df_electoral
```

```python
df_electoral.to_csv("electoral_week_comp.csv")
```
