```python
import pandas as pd
```

```python
most_recent=pd.read_csv("../data/mpox_august12.csv")
most_recent
```

```python
most_recent["Country_pure"]=most_recent["Country"].str.replace("*","")
most_recent
```

```python
previous_df=pd.read_csv("../data/mpox_june28.csv")
previous_df
```

```python
previous_df["Country_pure"]=previous_df["Country"].str.replace("*","")
previous_df
```

```python
df=pd.merge(most_recent,previous_df,on="Country_pure",suffixes=["_aug","_june"])
df
```

```python
df_diff=df[["Country_pure","Total Confirmed Cases_aug","Total Deaths_aug","Total Confirmed Cases_june","Total Deaths_june"]]
df_diff[df_diff.columns[1:]]=df_diff[df_diff.columns[1:]].astype(float)
df_diff
```

```python
df_diff["sixweeks_cases"]=df_diff["Total Confirmed Cases_aug"]-df_diff["Total Confirmed Cases_june"]
df_diff["sixweeks_death"]=df_diff["Total Deaths_aug"]-df_diff["Total Deaths_june"]
df_diff
```

```python
last_month_cases=df_diff[df_diff["sixweeks_cases"]>0]
last_month_cases
```

```python
last_month_cases.to_csv("../scripts/last_month_cases_here.csv")
```
