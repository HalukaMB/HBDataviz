```python
import pandas as pd
import numpy as np
```

```python
data2010=pd.read_csv("../data/2010before.csv")
data2010["Zuletzt"]=data2010["Zuletzt"].str.replace(".","")
data2010["Zuletzt"]=data2010["Zuletzt"].str.replace(",",".")
data2010
```

```python
data2024=pd.read_csv("../data/2024.csv")
data2024["Zuletzt"]=data2024["Zuletzt"].str.replace(".","")
data2024["Zuletzt"]=data2024["Zuletzt"].str.replace(",",".")
data2024
```

```python
WendeToNow=pd.concat([data2010,data2024],axis=0).drop_duplicates(subset=["Datum"])
WendeToNow
```

```python
WendeToNow["Datum"]=pd.to_datetime(WendeToNow["Datum"])
```

```python
WendeToNow=WendeToNow.sort_values(by=["Datum"]).reset_index(drop=True)
WendeToNow
```

```python
WendeToNow
```

```python
WendeToNow["28daysSmoothened"]=WendeToNow["Zuletzt"].rolling(window=28).mean()
WendeToNow["diffSmoothened"]=WendeToNow["28daysSmoothened"]-WendeToNow["28daysSmoothened"].shift(1)
WendeToNow
```

```python
WendeToNow[WendeToNow["Datum"]>"2023-01-01"]
```

```python
WendeToNow[WendeToNow["Datum"]>"2023-01-01"].to_csv("recent_dax.csv",index=False)
```

```python
negative_to_positve = (WendeToNow["diffSmoothened"] > 0) & (WendeToNow["diffSmoothened"].shift(-1) > 0)
positive_to_negative = (WendeToNow["diffSmoothened"] < 0) & (WendeToNow["diffSmoothened"].shift(-1) < 0)
WendeToNow["Trend"]=0
WendeToNow["Trend"]=np.where(positive_to_negative,-1,WendeToNow["Trend"])
WendeToNow["Trend"]=np.where(negative_to_positve,1,WendeToNow["Trend"])
WendeToNow
```

```python
listOfTrends=[]
trendcounter=0
for index, row in WendeToNow.iterrows():
    if (row["Trend"]==0):
        trendcounter=0
    else:
        trendcounter=trendcounter+row["Trend"]
    listOfTrends.append(trendcounter)

WendeToNow["daysOfTrend"]=listOfTrends
```
