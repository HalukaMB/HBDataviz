```python
import pandas as pd
```

```python
data=pd.read_csv("../data/GrippeWeb_Daten_des_Wochenberichts.tsv",sep="\t")
data
```

```python
all_ages=data[data["Altersgruppe"]=="00+"]
all_ages
```

```python
all_ages_bundesweit=all_ages[all_ages["Region"]=="Bundesweit"]
all_ages_bundesweit
```

```python
all_ages_bundesweit["kw"]=all_ages_bundesweit["Kalenderwoche"].str[-2:]
all_ages_bundesweit["year"]=all_ages_bundesweit["Kalenderwoche"].str[:4]

all_ages_bundesweit["yearkw"]=all_ages_bundesweit["year"]+all_ages_bundesweit["kw"]
```

```python
all_ages_bundesweit_are=all_ages_bundesweit[all_ages_bundesweit["Erkrankung"]=="ARE"]
all_ages_bundesweit_are
```

```python
pre_pandemic_are=all_ages_bundesweit_are[all_ages_bundesweit_are["year"].astype(int)<2020]
pre_pandemic_are
```

```python
post_pandemic_are=all_ages_bundesweit_are[(all_ages_bundesweit_are["year"].astype(int)>=2020) & (all_ages_bundesweit_are["year"].astype(int)<2024)]
post_pandemic_are
```

```python
pre_pandemic_are_max_levels=pre_pandemic_are[["kw","Inzidenz"]].groupby("kw").max().reset_index()
pre_pandemic_are_mean_levels=pre_pandemic_are[["kw","Inzidenz"]].groupby("kw").mean().reset_index()
```

```python
post_pandemic_are_max_levels=post_pandemic_are[["kw","Inzidenz"]].groupby("kw").max().reset_index()
post_pandemic_are_mean_levels=post_pandemic_are[["kw","Inzidenz"]].groupby("kw").mean().reset_index()
```

```python
pd.merge(post_pandemic_are_max_levels,post_pandemic_are_levels,on="kw",suffixes=["_pre","_post"])
```
