```python
bis_2008 = pd.read_csv("/content/drive/MyDrive/Fed/DFEDTAR.csv")
ab_2008 = pd.read_csv("/content/drive/MyDrive/Fed/DFEDTARU.csv")
```

```python
bis_2008.columns = ["date","target_rate"]
bis_2008.head()
```

```python
ab_2008.columns = ["date","target_rate"]
ab_2008.head()
```

```python
target_rates = pd.concat([bis_2008,ab_2008],ignore_index=True)
target_rates
```

```python
target_rates["date"] = pd.to_datetime(target_rates["date"], format = "%Y-%m-%d", errors="coerce")
```

```python
# Tage zwischen den Eintrgen in der date Spalte
target_rates["Tage_Differenz"] = target_rates["date"].diff().dt.days
target_rates
```

```python
target_rates["Tage_Differenz"].value_counts()
```

```python
target_rates["diff"] = target_rates["target_rate"].diff()
target_rates['vernderung'] = target_rates['diff'].apply(lambda x: 'gestiegen' if x > 0 else ('gesunken' if x < 0 else 'gleichgeblieben'))
target_rates.head()
```

```python
target_rates['Tage_ohne_Vernderung'] = target_rates.groupby((target_rates['vernderung'] != target_rates['vernderung'].shift()).cumsum())
```

```python
target_rates
```

```python
target_rates.sort_values(by="Tage_ohne_Vernderung",ascending=False)
```

```python
target_rates["keine_vernderung"] = target_rates["diff"] ==0
```

```python
target_rates
```

```python
target_rates['Tage_seit_letzter_Vernderung'] = (target_rates['keine_vernderung'] != target_rates['keine_vernderung'].shift()).cumsum()
```

```python
zeitrume_ohne_vernderung = target_rates[target_rates['keine_vernderung']]
lnge_zeitrume = zeitrume_ohne_vernderung.groupby('Tage_seit_letzter_Vernderung').agg(
    start_date=('date', 'first'),
    end_date=('date', 'last'),
    length=('date', 'size'),
    zielrate = ("target_rate", "first")
).reset_index(drop=True)
```

**Prsidentschaftswahljahre**

```python
import numpy as np
from datetime import datetime, timedelta
```

```python
def erster_dienstag_nach_erstem_november(jahr):
  november_1 = datetime(jahr, 11,1)
  if november_1.weekday() <=1:
    return november_1 + timedelta(days=(1 - november_1.weekday() +7)%7)
  else:
    return november_1 + timedelta(days=(8 - november_1.weekday()+1)%7)


wahljahre = []
for jahr in range(1984, 2025, 4):
  start_date = datetime(jahr, 1, 1)
  end_date = erster_dienstag_nach_erstem_november(jahr)
  wahljahre.append((start_date, end_date))


zins_wahljahre = pd.DataFrame()

for start_date, end_date in wahljahre:
  wahlen = target_rates[(target_rates["date"] >= start_date) & (target_rates["date"] <= end_date)]
  zins_wahljahre = pd.concat([zins_wahljahre, wahlen])

zins_wahljahre.reset_index(drop=True, inplace = True)
```
