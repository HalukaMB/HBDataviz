```python
import pandas as pd
```

```python
data=pd.read_excel("data/Lebensmittelkontrollen.xlsx")
data[data["Stadt"]=="Bonn?"]
data["Stadt"].unique()
```

```python
data[data["Stadt"]=="Berlin"]
```

```python
data[data["Stadt"]=="Dresden?"]
```

```python
data_pivot=pd.pivot_table(data, values='Quote', index=['Stadt'],
                       columns=['Jahr'])
data_pivot=data_pivot.reset_index()
data_pivot
```

```python
data_pivot["Stadt"]=data_pivot["Stadt"].str.replace("?","")
data_pivot
```

```python
data_pivot_red=data_pivot[["Stadt",2021,2022,2023]]
data_pivot_red
```

```python
data_pivot_red[[2021,2022,2023]]=data_pivot_red[[2021,2022,2023]]*100
data_pivot_red["mean"]=data_pivot_red[[2021,2022,2023]].mean(axis=1)
data_pivot_red
```

```python
data_pivot_worst=data_pivot_red.sort_values(by="mean",ascending=False).head()
data_pivot_worst
```

```python
data_pivot_worst.to_csv("staedte_ranking_worst.csv",index=False)
```

```python
data_pivot_best=data_pivot_red.sort_values(by="mean",ascending=True).head()
data_pivot_best
```

```python
data_pivot_best.to_csv("staedte_ranking_best.csv",index=False)
```

```python
data_pivot_rest=data_pivot_worst.iloc[6:-5]
data_pivot_rest.to_csv("staedte_ranking_rest.csv",index=False)
```

```python
data_pivot_rest
```

```python
rest=pd.DataFrame({"Stadt":["Münster","Duisburg"]})
```

```python
pd.concat([data_pivot_rest,rest]).to_csv("staedte_ranking_rest.csv",index=False)
```
