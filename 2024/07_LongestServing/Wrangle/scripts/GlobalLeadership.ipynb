```python
import numpy as np
```

```python
executives["terms"]=0
executives
```

```python
executives["terms"]=np.where(executives.person_office_type_1=="Executive",executives["person_office_terms_1"],executives["terms"])
executives["terms"]=np.where(executives.person_office_type_2=="Executive",executives["person_office_terms_2"],executives["terms"])
executives["terms"]=np.where(executives.person_office_type_3=="Executive",executives["person_office_terms_3"],executives["terms"])
executives["terms"]
```

```python
executives["startReign"]="2024-03-13"
executives["startReign"]=np.where(executives.person_office_type_1=="Executive",executives["person_office_service_began_1"],executives["startReign"])
executives["startReign"]=np.where(executives.person_office_type_2=="Executive",executives["person_office_service_began_2"],executives["startReign"])
executives["startReign"]=np.where(executives.person_office_type_3=="Executive",executives["person_office_service_began_3"],executives["startReign"])
executives["startReign"]
```

```python
executives_by_terms=executives.sort_values(by=["terms"],ascending=False).head(200)

executives_by_terms.to_csv("executives_by_terms.csv", index=False)
executives_by_terms=executives_by_terms[executives_by_terms["terms"]>1]
executives_by_terms
```

```python
executives_by_reign=executives.sort_values(by="startReign",ascending=False)
executives_by_reign=executives[executives["startReign"]<"2014-01-01"]

executives_by_reign.sort_values(by="startReign",ascending=False)
```

```python
executives_long_career=pd.concat([executives_by_terms,executives_by_reign])
executives_long_career
```

```python
executives_long_career_without_dup=executives_long_career.drop_duplicates(subset="person_nid")
```
