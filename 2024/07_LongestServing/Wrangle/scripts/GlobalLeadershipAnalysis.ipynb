```python
worldbank=pd.read_csv("../data/a5f7d039-d9dc-4ed1-8e77-8ecff8f27793_Data.csv")
gdp_per_cap=worldbank[worldbank["Series Name"]=="GDP per capita (current US$)"]
gdp_per_cap
```

```python
gdp_per_cap["GermanName"]=gdp_per_cap["Country Name"].replace(reverse_country_map)
gdp_per_cap
```

```python
reign_20_years=reign_20_years.sort_values("duration",ascending=False)
```

```python
gpd_concat=pd.DataFrame()
gpd_concat_rel=pd.DataFrame()
for index, row in reign_20_years.iterrows():
    countryname=row["country_title"]
    year=(row["YEAR"])
    country_gdp=gdp_per_cap[gdp_per_cap["GermanName"]==countryname]
    #country_gdp.columns.astype(str).contains(year)
    match_cols = [col for col in country_gdp.columns if str(year) in col]
    fittingno=country_gdp.columns.get_loc(match_cols[0])
    country_year_start=country_gdp.iloc[:,fittingno:]
    country_year_start["year"]=year
    gpd_concat=pd.concat([gpd_concat,country_year_start])
    while math.isnan(pd.to_numeric(country_gdp.iloc[0,fittingno],errors="coerce")):
        fittingno=fittingno+1
    Series=pd.to_numeric(country_gdp.iloc[0,fittingno:-1],errors="coerce")/pd.to_numeric(country_gdp.iloc[0,fittingno],errors="coerce")
    country_gpd_rel=pd.DataFrame({countryname:Series})
    gpd_concat_rel=pd.concat([gpd_concat_rel,country_gpd_rel],axis=1)
```

```python
perc_100_gdp=gpd_concat_rel*100-100
perc_100_gdp
```

```python
gpd_concat.to_csv("gpd.csv",index=False)
```

```python
perc_100_gdp.index=perc_100_gdp.index.astype(str).str[:4]
perc_100_gdp
```
