```python
xl = pd.read_excel('../data/Fiscal timebomb.xlsx',sheet_name=None);

sheet_names=pd.DataFrame([xl]).columns
sheet_names.tolist()
```

```python
df = pd.read_excel('../data/Fiscal timebomb.xlsx',sheet_name="Germany");
overall_df=df[(df["Indicator"]=="Debt (% GDP)")&(df["Source"]=="IMF")]
overall_df=overall_df[["indicator_date","Value"]]
overall_df
for sheet in sheet_names.tolist()[1:]:
    df = pd.read_excel('../data/Fiscal timebomb.xlsx',sheet_name=sheet);
    try:
        reduced_df=df[(df["Indicator"]=="Debt (% GDP)")&(df["Source"]=="IMF")]
        reduced_df=reduced_df.reset_index(drop=True)
        reduced_df[sheet]=reduced_df["Value"]
        reduced_df=reduced_df[["indicator_date",sheet]]
        overall_df=pd.merge(overall_df,reduced_df,on="indicator_date",how="outer")
    except:
        pass
overall_df
```

```python
span_20_years=overall_df[(overall_df["indicator_date"]<2035)&(overall_df["indicator_date"]>2014)]
span_20_years
```

```python
span_20_years=span_20_years.dropna(axis=1)
span_20_years
```

```python
df = pd.read_excel('../data/Fiscal timebomb.xlsx',sheet_name="China");
df.Source.unique()
```

```python
df = pd.read_excel('../data/Fiscal timebomb.xlsx',sheet_name="Germany");
overall_df_bloomberg=df[(df["Indicator"]=="Debt (% GDP)")&(df["Source"]=="Bloomberg Economics")]
overall_df_bloomberg=overall_df_bloomberg[["indicator_date","Value"]]
overall_df_bloomberg
for sheet in sheet_names.tolist()[1:]:
    df = pd.read_excel('../data/Fiscal timebomb.xlsx',sheet_name=sheet);
    try:
        reduced_df=df[(df["Indicator"]=="Debt (% GDP)")&(df["Source"]=="Bloomberg Economics")]
        reduced_df=reduced_df.reset_index(drop=True)
        reduced_df[sheet]=reduced_df["Value"]
        reduced_df=reduced_df[["indicator_date",sheet]]
        overall_df_bloomberg=pd.merge(overall_df_bloomberg,reduced_df,on="indicator_date",how="outer")
    except:
        pass
overall_df_bloomberg
```

```python
span_20_years_bloomberg=overall_df_bloomberg[(overall_df_bloomberg["indicator_date"]<2035)&(overall_df_bloomberg["indicator_date"]>2014)]
span_20_years_bloomberg=span_20_years_bloomberg.dropna(axis=1)
```

```python
imf_data=span_20_years[span_20_years.columns[span_20_years.columns!="Value"]]
imf_data
```

```python
additions_from_bloomberg=[]
for entry in span_20_years_bloomberg:
    if(entry not in(imf_data.columns.tolist())):
        additions_from_bloomberg.append(entry)
        
```

```python
span_20_years_bloomberg_addition=span_20_years_bloomberg[additions_from_bloomberg]
```

```python
imf_bloomberg=pd.concat([imf_data,span_20_years_bloomberg_addition],axis=1)
imf_bloomberg=imf_bloomberg.reset_index(drop=True)
imf_bloomberg
```
