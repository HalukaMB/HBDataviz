```python
import geopandas as gpd
```

```python
file=gpd.read_file("../data/EPSG_25832/VG250_GEM.shp")
```

```python
file
```

```python
no_dup_polygons=file.drop_duplicates(subset=["AGS"],keep="first").sort_values(by="AGS")
no_dup_polygons
```

```python
duplicated_polygons=file[file.duplicated(subset="AGS",keep=False)].sort_values(by="AGS")
duplicated_polygons
```

```python
import shapely
```

```python
duplicate_ags=duplicated_polygons["AGS"].unique()
```

```python
duplicate_geometries = duplicated_polygons.dissolve(by='AGS', aggfunc='sum').reset_index()
duplicate_geometries=duplicate_geometries[["AGS","geometry"]]
```

```python
names_of_duplicates=duplicated_polygons[["AGS","GEN"]].drop_duplicates()
```

```python
duplicates_summed = duplicate_geometries.merge(names_of_duplicates, on='AGS')
duplicates_summed=duplicates_summed[["AGS","GEN","geometry"]]
duplicates_summed
```

```python
no_dup_polygons=no_dup_polygons[["AGS","GEN","geometry"]]
no_dup_polygons
```

```python
concat_dup=pd.concat([no_dup_polygons,duplicates_summed])
concat_dup.drop_duplicates(subset=["AGS"],keep="last")
concat_dup.to_file("summed_ags.geojson")
```

```python
nord=concat_dup[concat_dup["AGS"].str[:2].isin(["01","02","03","04","13"])]
west=concat_dup[concat_dup["AGS"].str[:2].isin(["05","06","07","10"])]
ost=concat_dup[concat_dup["AGS"].str[:2].isin(["11","12","15","16"])]
sued=concat_dup[concat_dup["AGS"].str[:2].isin(["08","09"])]
```

```python
nord.to_crs(4326).to_file("nord.geojson")
west.to_crs(4326).to_file("west.geojson")
ost.to_crs(4326).to_file("ost.geojson")
sued.to_crs(4326).to_file("sued.geojson")
```
