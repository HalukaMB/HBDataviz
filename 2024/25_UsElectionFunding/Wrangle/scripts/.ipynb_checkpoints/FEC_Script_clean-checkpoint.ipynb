## This allows us to go into filings of the primary committees of candidates, month by month. Does not work for pacs

```python
harris_committee_dictionary={ "HARRIS FOR PRESIDENT":"C00703975",
    "FUTURE FORWARD USA":"C00669259",
    "THE LINCOLN PROJECT":"C00725820"
}
```

```python
trump_committee_dictionary={"DONALD J. TRUMP FOR PRESIDENT 2024":"C00828541",
 "MAKE AMERICA GREAT AGAIN":"C00580100",
    "SAVE AMERICA PAC":"C00461723"
}
```

```python
key=PERSONAL_KEY
```

```python
dictionary=harris_committee_dictionary
def get_monthly_receipts(dictionary):
    results_dict={}
    for name in list(dictionary.keys()):
        committee_id=dictionary[name]
        url_reports=(f"https://api.open.fec.gov/v1/reports/presidential/?page=1&per_page=100&cycle=2022&cycle=2024&committee_id={committee_id}"
                    f"&sort=-coverage_end_date&sort_hide_null=false&sort_null_only=false&sort_nulls_last=false&api_key={key}")
        r = requests.get(url_reports)
        data = r.json()
        data_dict={}
        for element in (data["results"]):
            print(element["coverage_end_date"])
            print(element["total_receipts_ytd"])
            data_dict[element["coverage_end_date"]]=element["total_receipts_ytd"]
        results_dict[name]=data_dict
    return(results_dict)
```

```python
harris_receipts=get_monthly_receipts(harris_committee_dictionary)
harris_receipts
```

```python
trump_receipts=get_monthly_receipts(trump_committee_dictionary)
trump_receipts
```

```python
trump_df=pd.DataFrame(trump_receipts["DONALD J. TRUMP FOR PRESIDENT 2024"],index=["Trump"]).T
trump_df
```

```python
harris_df=pd.DataFrame(harris_receipts["HARRIS FOR PRESIDENT"],index=["Harris"]).T
harris_df
```

```python
comparison_trump_harris=pd.concat([harris_df,trump_df],axis=1)
comparison_trump_harris.index=comparison_trump_harris.index.set_names(['date'])
comparison_trump_harris=comparison_trump_harris.reset_index()
comparison_trump_harris
```

```python
comparison_trump_harris_last_two_years=comparison_trump_harris[comparison_trump_harris["date"]>="2022-12-31T00:00:00"].to_csv("trump_harris.csv",index=False)
comparison_trump_harris_last_two_years
```

## Also we can ask for a breakdown of donations by size

```python
def contribution_split(candidate_id):
    url_contribution=(f"https://api.open.fec.gov/v1/presidential/contributions/by_size/?page=1&per_page=20&election_year=2024&candidate_id={candidate_id}"
    f"&sort=size&sort_hide_null=false&sort_null_only=false&sort_nulls_last=false&api_key={key}")
    r = requests.get(url_contribution)
    data = r.json()
    return data["results"]
```

```python
trump_contributions=contribution_split(trump_id)
harris_contributions=contribution_split(harris_id)
harris_contributions
```

```python
trump_contributions=pd.DataFrame(trump_contributions)
```

```python
harris_contributions=pd.DataFrame(harris_contributions)
```

```python
size_comp=pd.merge(trump_contributions,harris_contributions,on=["election_year","size","size_range_id"],suffixes=["_Trump","_Harris"])
size_comp_red=size_comp[["size","contribution_receipt_amount_Trump","contribution_receipt_amount_Harris"]]
size_comp_red
```

```python
size_comp_red["size_groups"]="no class"
size_comp_red["size_groups"]=np.where(size_comp_red["size"]==0,"1-200$",size_comp_red["size_groups"])
size_comp_red["size_groups"]=np.where(size_comp_red["size"]==200,"200-500$",size_comp_red["size_groups"])
size_comp_red["size_groups"]=np.where(size_comp_red["size"]==500,"500-1000$",size_comp_red["size_groups"])
size_comp_red["size_groups"]=np.where(size_comp_red["size"]==1000,"1000-2000$",size_comp_red["size_groups"])
size_comp_red["size_groups"]=np.where(size_comp_red["size"]==2000,"über 2000$",size_comp_red["size_groups"])
size_comp_red
```

```python
size_comp_red=size_comp_red[["size_groups","contribution_receipt_amount_Trump","contribution_receipt_amount_Harris"]]
size_comp_red.columns=["Spendengröße","Trump","Harris"]
size_comp_red
```

```python
size_comp_red.to_csv("contributions_size.csv",index=False)
```

## But we can use a different endpoint to get data for other committees

```python
republican_dict={"Republican National Committee":"C00003418",
                 "National Republican Senatorial Committee":"C00027466",
                 "National Republican Congressional Committee":"C00075820", "MAKE AMERICA GREAT AGAIN":"C00580100",
    "SAVE AMERICA PAC":"C00461723"}
democratic_dict={
                "Democratic Senatorial Campaign Committee":"C00042366",
    "Democratic National Committee":"C00010603",
                "Democratic Congressional Campaign Committee":"C00000935",
    "FUTURE FORWARD USA":"C00669259",
    "THE LINCOLN PROJECT":"C00725820"}
```

```python
def get_monthly_receipts_partycommittees(dictionary):
    results_dict={}
    for name in list(dictionary.keys()):
        print(name)
        committee_id=dictionary[name]
        url_reports=(f"https://api.open.fec.gov/v1/committee/{committee_id}/reports/?page=1&per_page=100"
         f"&sort=-coverage_end_date&sort_hide_null=false&sort_null_only=false&sort_nulls_last=false&api_key={key}")
        r = requests.get(url_reports)
        data = r.json()
        data_dict={}
        for element in (data["results"]):
            #print(element["coverage_end_date"])
            #print(element["total_receipts_ytd"])
            data_dict[element["coverage_end_date"]]=element["total_receipts_ytd"]
        results_dict[name]=data_dict
    return(results_dict)
```

```python
democratic_receipts=get_monthly_receipts_partycommittees(democratic_dict)
republican_receipts=get_monthly_receipts_partycommittees(republican_dict)
```

```python
democratic_df=pd.DataFrame.from_dict(democratic_receipts,orient='index').T
republican_df=pd.DataFrame.from_dict(republican_receipts,orient='index').T
democratic_df.index=democratic_df.index.set_names(['date'])
republican_df.index=republican_df.index.set_names(['date'])
democratic_df=democratic_df.reset_index()
republican_df=republican_df.reset_index()
democratic_df=democratic_df.sort_values(by="date")
republican_df=republican_df.sort_values(by="date")
```

```python
republican_df[republican_df.columns[1:]]=republican_df[republican_df.columns[1:]].astype(float)
democratic_df[democratic_df.columns[1:]]=democratic_df[democratic_df.columns[1:]].astype(float)
democratic_df
```
